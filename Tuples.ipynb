# 6.1 Tuples

## Tuple
Tuple digunakan untuk menyimpan beberapa item dalam satu variabel.

Tuple adalah salah satu dari 4 tipe data bawaan dalam Python yang digunakan untuk menyimpan kumpulan data, 3 lainnya adalah List , Set , dan Dictionary , semuanya dengan kualitas dan penggunaan yang berbeda.

Tuple adalah suatu koleksi yang teratur dan tidak dapat diubah .

Tuple ditulis dengan tanda kurung bulat.

```python
#Contoh
#Buat Tuple:

thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

## Item Tuple
Item tupel bersifat teratur, tidak dapat diubah, dan memperbolehkan nilai duplikat.

Item tuple diindeks, item pertama memiliki indeks [0], item kedua memiliki indeks, [1]dan seterusnya.

## Dipesan
Ketika kita katakan bahwa tupel diurutkan, artinya item-itemnya memiliki urutan yang pasti, dan urutan itu tidak akan berubah.

## Tidak dapat diubah
Tuple tidak dapat diubah, artinya kita tidak dapat mengubah, menambah atau menghapus item setelah tuple dibuat.

## Izinkan Duplikat
Karena tupel diindeks, tupel tersebut dapat memiliki item dengan nilai yang sama:

```python
#Contoh
#Tuple memperbolehkan nilai duplikat:

thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
```

## Panjang Tuple
Untuk menentukan berapa banyak item yang dimiliki suatu tupel, gunakan len()fungsi:

```python
#Contoh
#Cetak jumlah item dalam tupel:

thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
```

## Buat Tuple Dengan Satu Item
Untuk membuat tupel dengan hanya satu item, Anda harus menambahkan koma setelah item, jika tidak, Python tidak akan mengenalinya sebagai tupel.

```python
#Contoh
#Satu item tuple, ingat koma:

thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

## Item Tuple - Tipe Data
Item tupel dapat berupa tipe da|ta apa pun:

```python
#Contoh
#Tipe data string, int dan boolean:

tuple1 = ("apple", "banana", "cherry")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
```

Suatu tupel dapat berisi beberapa tipe data:

```python
#Contoh
#Tuple dengan string, integer, dan nilai boolean:

tuple1 = ("abc", 34, True, 40, "male")
```

## jenis()
Dari perspektif Python, tupel didefinisikan sebagai objek dengan tipe data 'tupel':

<class 'tuple'>

```python
#Contoh
#Apa tipe data suatu tupel?

mytuple = ("apple", "banana", "cherry")
print(type(mytuple))
```

## Konstruktor tuple()
Dimungkinkan juga untuk menggunakan konstruktor tuple() untuk membuat tuple.

```python
#Contoh
#Menggunakan metode tuple() untuk membuat tuple:

thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(thistuple)
```

## Akses Item Tuple
Anda dapat mengakses item tupel dengan merujuk ke nomor indeks, di dalam tanda kurung siku:

```python
#Contoh
#Cetak item kedua dalam tupel:

thistuple = ("apple", "banana", "cherry")
print(thistuple[1])
```

Catatan: Item pertama memiliki indeks 0.

## Pengindeksan Negatif
Pengindeksan negatif berarti mulai dari akhir.

-1merujuk pada item terakhir, -2merujuk pada item kedua terakhir, dst.

```python
#Contoh
#Cetak item terakhir dari tupel:

thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])
```

## Rentang Indeks
Anda dapat menentukan rentang indeks dengan menentukan di mana memulai dan di mana mengakhiri rentang.

Saat menentukan rentang, nilai yang dikembalikan akan berupa tupel baru dengan item yang ditentukan.

```python
#Contoh
#Kembalikan item ketiga, keempat, dan kelima:

thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])
```

Catatan: Pencarian akan dimulai pada indeks 2 (termasuk) dan berakhir pada indeks 5 (tidak termasuk).

Ingat bahwa item pertama memiliki indeks 0.

Dengan menghilangkan nilai awal, rentang akan dimulai pada item pertama:

```python
#Contoh
#Contoh ini mengembalikan item dari awal ke, tetapi TIDAK termasuk, "kiwi":

thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[:4])
```

Dengan menghilangkan nilai akhir, rentang akan berlanjut hingga akhir tupel:

```python
#Contoh
#Contoh ini mengembalikan item dari "cherry" dan ke akhir:

thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:])
```

## Rentang Indeks Negatif
Tentukan indeks negatif jika Anda ingin memulai pencarian dari akhir tupel:

```python
#Contoh
#Contoh ini mengembalikan item dari indeks -4 (termasuk) ke indeks -1 (dikecualikan)

thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[-4:-1])
```

## Periksa apakah Item Ada
Untuk menentukan apakah item tertentu ada dalam suatu tupel, gunakan kata inkunci:

```python
#Contoh
#Periksa apakah "apple" ada dalam tuple:

thistuple = ("apple", "banana", "cherry")
if "apple" in thistuple:
  print("Yes, 'apple' is in the fruits tuple")
```

Tuple tidak dapat diubah, artinya Anda tidak dapat mengubah, menambah, atau menghapus item setelah tuple dibuat.

Namun ada beberapa solusi.

## Ubah Nilai Tuple
Setelah sebuah tupel dibuat, Anda tidak dapat mengubah nilainya. Tupel tidak dapat diubah , atau disebut juga immutable .

Namun, ada solusinya. Anda dapat mengonversi tupel menjadi daftar, mengubah daftar, dan mengonversi daftar kembali menjadi tupel.

```python
#Contoh
#Ubah tupel menjadi daftar agar dapat mengubahnya:

x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

## Tambahkan Item
Karena tupel tidak dapat diubah, tupel tidak memiliki metode bawaan append(), tetapi ada cara lain untuk menambahkan item ke tupel.

1. Ubah menjadi daftar : Sama seperti solusi untuk mengubah tupel, Anda dapat mengubahnya menjadi daftar, menambahkan item Anda, dan mengubahnya kembali menjadi tupel.

```python
#Contoh
#Ubah tupel menjadi daftar, tambahkan "orange", dan ubah kembali menjadi tupel:

thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
```

2. Tambahkan tuple ke tuple . Anda diperbolehkan menambahkan tuple ke tuple. Jadi, jika Anda ingin menambahkan satu item (atau banyak), buatlah tuple baru dengan item tersebut, lalu tambahkan ke tuple yang sudah ada:

```python
#Contoh
#Buat tupel baru dengan nilai "orange", dan tambahkan tupel tersebut:

thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y

print(thistuple)
```

Catatan: Saat membuat tupel dengan hanya satu item, ingatlah untuk menyertakan koma setelah item, jika tidak, item tersebut tidak akan diidentifikasi sebagai tupel.

## Hapus Item
Catatan: Anda tidak dapat menghapus item dalam tupel.

Tuple tidak dapat diubah , jadi Anda tidak dapat menghapus item darinya, tetapi Anda dapat menggunakan solusi yang sama seperti yang kami gunakan untuk mengubah dan menambahkan item tuple:

```python
#Contoh
#Ubah tupel menjadi daftar, hapus "apple", dan ubah kembali menjadi tupel:

thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
```

Atau Anda dapat menghapus tupel sepenuhnya:

```python
#Contoh
#Kata delkunci dapat menghapus tupel sepenuhnya:

thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple) #this will raise an error because the tuple no longer exists
```

## Membongkar Tuple
Saat kita membuat sebuah tupel, biasanya kita menetapkan nilai padanya. Ini disebut "mengemas" sebuah tupel:

Contoh

Mengemas suatu tupel:

fruits = ("apple", "banana", "cherry")

Namun, dalam Python, kita juga dapat mengekstrak nilai kembali ke dalam variabel. Ini disebut "unpacking":

```python
#Contoh
#Membongkar tupel:

fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

Catatan: Jumlah variabel harus sesuai dengan jumlah nilai dalam tupel, jika tidak, Anda harus menggunakan tanda bintang untuk mengumpulkan nilai yang tersisa sebagai daftar.

## Menggunakan Asterisk*
Jika jumlah variabel lebih kecil dari jumlah nilai, Anda dapat menambahkan tanda *pada nama variabel dan nilai akan ditetapkan ke variabel tersebut sebagai daftar:

```python
#Contoh
#Tetapkan sisa nilai sebagai daftar yang disebut "merah":

fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)
```

Jika tanda bintang ditambahkan ke nama variabel lain selain yang terakhir, Python akan menetapkan nilai ke variabel tersebut hingga jumlah nilai yang tersisa cocok dengan jumlah variabel yang tersisa.

```python
#Contoh
#Tambahkan daftar nilai variabel "tropic":

fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

## Melakukan Loop Melalui Tuple
Anda dapat melakukan pengulangan melalui item tupel dengan menggunakan forperulangan.

```python
#Contoh
#Ulangi melalui item dan cetak nilainya:

thistuple = ("apple", "banana", "cherry")
for x in thistuple:
  print(x)
```

Pelajari lebih lanjut tentang forloop di Bab Python For Loops kami.

## Ulangi Melalui Angka Indeks
Anda juga dapat melakukan pengulangan melalui item tupel dengan merujuk pada nomor indeksnya.

Gunakan fungsi range()dan len()untuk membuat iterasi yang sesuai.

```python
#Contoh
#Cetak semua item dengan merujuk pada nomor indeksnya:

thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
  print(thistuple[i])
```

## Menggunakan Perulangan While
Anda dapat melakukan pengulangan melalui item tupel dengan menggunakan whileperulangan.

Gunakan len()fungsi tersebut untuk menentukan panjang tupel, lalu mulai dari 0 dan lakukan pengulangan melalui item tupel dengan merujuk pada indeksnya.

Ingatlah untuk menambah indeks sebesar 1 setelah setiap iterasi.

```python
#Contoh
#Cetak semua item, gunakan whileloop untuk menelusuri semua nomor indeks:

thistuple = ("apple", "banana", "cherry")
i = 0
while i < len(thistuple):
  print(thistuple[i])
  i = i + 1
```

## Gabungkan Dua Tuple
Untuk menggabungkan dua atau lebih tupel, Anda dapat menggunakan + operator:

```python
#Contoh
#Gabungkan dua tupel:

tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

## Kalikan Tuple
Jika Anda ingin mengalikan isi suatu tupel sejumlah tertentu kali, Anda dapat menggunakan * operator:

```python
#Contoh
#Kalikan tupel buah dengan 2:

fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(mytuple)
```
