#### 13.Daftar Tertaut
Struktur data linear yang menyimpan data dalam node, yang dihubungkan oleh pointer. Tidak seperti array, node dalam linked list tidak disimpan dalam lokasi memori yang bersebelahan dan hanya dapat diakses secara berurutan , dimulai dari awal list.
#### 14.Tumpukan
Struktur data linear yang mengikuti prinsip Last In, First Out (LIFO) . Tumpukan memainkan peran penting dalam mengelola pemanggilan fungsi, memori, dan banyak digunakan dalam algoritma seperti masalah rentang stok, elemen terbesar berikutnya, dan area terbesar dalam histogram.
#### 15.Antrean
Antrean adalah struktur data linier yang mengikuti prinsip First In, First Out (FIFO) . Antrean memainkan peran penting dalam mengelola tugas atau data dalam sistem pemesanan, penjadwalan, dan penanganan pesan.
#### 16.Deque
Deque atau antrean ujung ganda adalah struktur data yang memungkinkan elemen ditambahkan atau dihapus dari kedua ujung secara efisien.
#### 17.Pohon
Struktur data hierarkis non-linier yang terdiri dari simpul-simpul yang terhubung oleh tepi, dengan simpul teratas yang disebut akar dan simpul-simpul yang memiliki simpul anak. Struktur ini banyak digunakan dalam sistem berkas , basis data , algoritma pengambilan keputusan , dll.
#### 18.Tumpukan
Pohon biner lengkap yang memenuhi properti heap . Heap biasanya digunakan untuk mengimplementasikan antrean prioritas , di mana elemen terkecil atau terbesar selalu berada di akar pohon.
#### 19.Grafik
Struktur data non-linier yang terdiri dari himpunan titik (atau simpul) berhingga dan himpunan sisi (atau tautan) yang menghubungkan sepasang simpul. Graf banyak digunakan untuk merepresentasikan hubungan antar entitas.
#### 20.Algoritma Greedy
Algoritma Greedy membangun solusi satu per satu dan memilih solusi berikutnya yang memberikan manfaat paling nyata dan langsung, yaitu pilihan paling optimal saat itu. Oleh karena itu, permasalahan di mana pemilihan optimal lokal juga menghasilkan solusi global adalah yang paling sesuai untuk Greedy.

Consider the following program that attempts to locate an element x in a sorted array a[ ] using binary search. Assume N>1. The program is erroneous. Under what conditions does the program fail?

```python
def find(a, n, x):
    i = 0
    j = n
    while i < j:
        k = (i + j) // 2
        if a[k] < x:
            i = k + 1
        else:
            j = k
    
    if i < len(a) and a[i] == x:
        print("x is in the array")
    else:
        print("x is not in the array")
```
