# 7.2 Rekursi
Rekursi terjadi ketika sebuah fungsi memanggil dirinya sendiri.

Rekursi adalah konsep matematika dan pemrograman yang umum. Artinya, sebuah fungsi memanggil dirinya sendiri. Hal ini memiliki keuntungan karena memungkinkan kita untuk melakukan perulangan melalui data untuk mencapai suatu hasil.

Pengembang harus sangat berhati-hati dengan rekursi karena sangat mudah untuk tergelincir ke dalam penulisan fungsi yang tidak pernah berakhir, atau fungsi yang menggunakan memori atau daya prosesor secara berlebihan. Namun, jika ditulis dengan benar, rekursi dapat menjadi pendekatan pemrograman yang sangat efisien dan elegan secara matematis.

```python
#Contoh
#Sebuah fungsi rekursif sederhana yang menghitung mundur dari 5:

def countdown(n):
  if n <= 0:
    print("Done!")
  else:
    print(n)
    countdown(n - 1)

countdown(5)
```

## Kasus Dasar dan Kasus Rekursif
Setiap fungsi rekursif harus memiliki dua bagian:

- Kasus dasar - Suatu kondisi yang menghentikan rekursi
- Kasus rekursif - Fungsi memanggil dirinya sendiri dengan argumen yang dimodifikasi

Tanpa kasus dasar, fungsi tersebut akan memanggil dirinya sendiri tanpa henti, menyebabkan kesalahan stack overflow.

```python
#Contoh
#Mengidentifikasi kasus dasar dan kasus rekursif:

def factorial(n):
  # Base case
  if n == 0 or n == 1:
    return 1
  # Recursive case
  else:
    return n * factorial(n - 1)

print(factorial(5))
```

Kasus dasar sangat penting. Selalu pastikan fungsi rekursif Anda memiliki kondisi yang pada akhirnya akan terpenuhi.

### Deret Fibonacci
Deret Fibonacci adalah contoh klasik di mana setiap angka merupakan jumlah dari dua angka sebelumnya. Deret ini dimulai dengan 0 dan 1:

0, 1, 1, 2, 3, 5, 8, 13, ...

Urutan tersebut berlanjut tanpa batas, dengan setiap angka merupakan jumlah dari dua angka sebelumnya.

Kita dapat menggunakan rekursi untuk menemukan angka tertentu dalam urutan tersebut:

```python
#Contoh
#Temukan angka ke-7 dalam deret Fibonacci:

def fibonacci(n):
  if n <= 1:
    return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(7))
```

## Rekursi dengan Daftar
Rekursi dapat digunakan untuk memproses daftar dengan menangani satu elemen pada satu waktu:

```python
#Contoh
#Hitung jumlah semua elemen dalam sebuah daftar:

def sum_list(numbers):
  if len(numbers) == 0:
    return 0
  else:
    return numbers[0] + sum_list(numbers[1:])

my_list = [1, 2, 3, 4, 5]
print(sum_list(my_list))
```

```python
#Contoh
#Temukan nilai maksimum dalam sebuah daftar:

def find_max(numbers):
  if len(numbers) == 1:
    return numbers[0]
  else:
    max_of_rest = find_max(numbers[1:])
    return numbers[0] if numbers[0] > max_of_rest else max_of_rest

my_list = [3, 7, 2, 9, 1]
print(find_max(my_list))
```

## Batasan Kedalaman Rekursi
Python memiliki batasan seberapa dalam rekursi dapat dilakukan. Batasan default biasanya sekitar 1000 panggilan rekursif.

## Fungsi Rekursif
Fungsi rekursif adalah fungsi yang memanggil dirinya sendiri untuk menyelesaikan suatu masalah. Fungsi ini umum digunakan dalam masalah matematika dan strategi bagi-dan-taklukkan. Selalu sertakan kasus dasar untuk menghindari rekursi tak terbatas.

```python
def factorial(n):
    if n == 0:  
        return 1
    else:
        return n * factorial(n - 1) 
      
print(factorial(4))
```

Di sini kita telah membuat fungsi rekursif untuk menghitung faktorial dari suatu bilangan. Fungsi ini memanggil dirinya sendiri hingga kondisi dasar (n==0) terpenuhi.

Rekursi adalah teknik pemrograman di mana sebuah fungsi memanggil dirinya sendiri secara langsung atau tidak langsung untuk menyelesaikan suatu masalah dengan memecahnya menjadi submasalah yang lebih kecil dan lebih sederhana.

Dalam Python, rekursi sangat berguna untuk masalah yang dapat dibagi menjadi tugas-tugas kecil yang identik, seperti perhitungan matematika, penelusuran pohon, atau algoritma bagi-dan-taklukkan.

Cara Kerja Rekursi
Fungsi rekursif sama seperti fungsi Python lainnya, kecuali bahwa fungsi tersebut memanggil dirinya sendiri di dalam badannya. Mari kita lihat struktur dasar fungsi rekursif:

def recursive_function(parameters): 

    if base_case_condition: 
    
        return base_result 
        
    else: 
    
        return recursive_function(modified_parameters)

Fungsi rekursif terdiri dari dua bagian utama:

- Kasus Dasar: Kondisi penghentian yang mencegah rekursi tak terbatas.
- Kasus Rekursif: Bagian dari fungsi di mana fungsi tersebut memanggil dirinya sendiri dengan parameter yang dimodifikasi.

Contoh-contoh Rekursi

Mari kita pahami rekursi dengan lebih baik melalui beberapa contoh.

Contoh 1: Perhitungan Faktorial

Kode ini mendefinisikan fungsi rekursif untuk menghitung faktorial suatu bilangan, di mana fungsi tersebut berulang kali memanggil dirinya sendiri dengan nilai yang lebih kecil hingga mencapai kasus dasar.

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:       # Recursive case
        return n * factorial(n - 1)

print(factorial(5))
```

Penjelasan:
- Kasus Dasar: Ketika n == 0 , rekursi berhenti dan mengembalikan 1 .
- Kasus Rekursif: Mengalikan n dengan faktorial dari n-1 hingga mencapai kasus dasar.

Contoh 2: Deret Fibonacci

Kode ini mendefinisikan fungsi rekursif untuk menghitung bilangan Fibonacci ke-n , di mana setiap bilangan adalah jumlah dari dua bilangan sebelumnya, dimulai dari 0 dan 1.

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10))
```

Penjelasan:

- Kasus Dasar: Jika n == 0, fungsi mengembalikan 0. Jika n == 1, fungsi mengembalikan 1. Kedua kasus ini diperlukan untuk menghentikan rekursi.
- Kasus Rekursif: fungsi memanggil dirinya sendiri dua kali dengan pengurangan n (yaitu, fibonacci(n-1) dan fibonacci(n-2)), menjumlahkan hasil dari panggilan-panggilan ini.

## Jenis-jenis Rekursi dalam Python
Rekursi secara umum dapat diklasifikasikan menjadi dua jenis: rekursi ekor dan rekursi non-ekor . Perbedaan utama antara keduanya berkaitan dengan apa yang terjadi setelah panggilan rekursif.

- Rekursi Ekor: Panggilan rekursif adalah hal terakhir yang dilakukan fungsi, jadi tidak ada yang terjadi setelah fungsi tersebut kembali. Beberapa bahasa pemrograman dapat mengoptimalkan ini agar berfungsi seperti perulangan, sehingga menghemat memori.
- Rekursi Non-Ekor: Fungsi tersebut melakukan lebih banyak pekerjaan setelah panggilan rekursif kembali, sehingga tidak dapat dioptimalkan menjadi sebuah loop.

Kode ini membandingkan rekursi ekor dan rekursi non-ekor menggunakan dua versi fungsi faktorial, satu dengan akumulator (rekursif ekor) dan satu dengan perkalian setelah panggilan rekursif (non-rekursif ekor).

```python
def tail_fact(n, acc=1):
    # Base case
    if n == 0:
        return acc
    # Tail recursive call with an accumulator
    else:
        return tail_fact(n-1, acc * n)

def nontail_fact(n):
    # Base case
    if n == 1:
        return 1
    # Non-tail recursive call because the multiplication happens after the call
    else:
        return n * nontail_fact(n-1)

# Example usage
print(tail_fact(5))  
print(nontail_fact(5))
```

Penjelasan:

- def tail_fact(n, acc=1): - Mendefinisikan fungsi faktorial rekursif ekor dengan akumulator acc untuk menyimpan hasil sementara.
- Jika n == 0: kembalikan acc - Kasus dasar: ketika n mencapai 0, kembalikan hasil akumulasi.
- return tail_fact(n-1, acc * n) - Panggilan rekursif ekor: mengalikan acc dengan n sebelum panggilan, sehingga tidak ada pekerjaan tambahan yang tersisa setelah rekursi.
- def nontail_fact(n): - Mendefinisikan fungsi faktorial non-rekursif ekor.
- Jika n == 1: kembalikan 1 - Kasus dasar: faktorial dari 1 adalah 1.
- return n * nontail_fact(n-1) - Panggilan non-ekor: perkalian terjadi setelah panggilan rekursif kembali, sehingga masih ada pekerjaan yang harus dilakukan setelah rekursi.
