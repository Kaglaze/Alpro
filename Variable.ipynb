## 1.3 Python Variables

Dalam Python, variabel digunakan untuk menyimpan data yang dapat direferensikan dan dimanipulasi selama eksekusi program. Variabel pada dasarnya adalah nama yang diberikan untuk suatu nilai. Tidak seperti banyak bahasa pemrograman lainnya, variabel Python tidak memerlukan deklarasi tipe secara eksplisit. Tipe variabel disimpulkan berdasarkan nilai yang diberikan.

Variabel berfungsi sebagai tempat penampung data. Variabel memungkinkan kita menyimpan dan menggunakan kembali nilai dalam program kita.

```python
# Variable 'x' stores the integer value 10
x = 5

# Variable 'name' stores the string "Samantha"
name = "Samantha"  

print(x)
print(name)
```

Dalam artikel ini, kita akan menjelajahi konsep variabel dalam Python, termasuk sintaksis, karakteristik, dan operasi umum.

### Rules for Naming Variables

Untuk menggunakan variabel secara efektif, kita harus mengikuti aturan penamaan Python:
+ Nama variabel hanya dapat berisi huruf, angka, dan garis bawah ( _).
+ Nama variabel tidak dapat dimulai dengan angka.
+ Nama variabel peka huruf besar-kecil ( myVardan myvarberbeda).
+ Hindari penggunaan kata kunci Python (misalnya, if, else, for) sebagai nama variabel.

### Assigning Values to Variables

#### Basic Assignment

Variabel dalam Python diberi nilai menggunakan = operator.

```python
x = 5
y = 3.14
z = "Hi"
```

#### Dynamic Typing

Variabel Python diketik secara dinamis, artinya variabel yang sama dapat menyimpan berbagai jenis nilai selama eksekusi.

```python
x = 10
x = "Now a string"
```

### Multiple Assignments

Python memperbolehkan beberapa variabel diberi nilai dalam satu baris.

#### Assigning the Same Value

Python memungkinkan penetapan nilai yang sama ke beberapa variabel dalam satu baris, yang dapat berguna untuk menginisialisasi variabel dengan nilai yang sama.

```python
a = b = c = 100
print(a, b, c)
```

#### Assigning Different Values

Kita dapat menetapkan nilai yang berbeda pada beberapa variabel secara bersamaan, membuat kode menjadi ringkas dan lebih mudah dibaca.

```python
x, y, z = 1, 2.5, "Python"
print(x, y, z)
```

### Type Casting a Variable

Pengecoran tipe mengacu pada proses konversi nilai dari satu tipe data ke tipe data lainnya. Python menyediakan beberapa fungsi bawaan untuk memfasilitasi pengecoran, termasuk int(), float(), dan str().

#### Basic Casting Functions

+ int() - Mengonversi nilai yang kompatibel menjadi integer.
+ float() - Mengubah nilai menjadi angka floating-point.
+ str() - Mengubah tipe data apa pun menjadi string.

#### Examples of Casting:

```python
# Casting variables
s = "10"  # Initially a string
n = int(s)  # Cast string to integer
cnt = 5
f = float(cnt)  # Cast integer to float
age = 25
s2 = str(age)  # Cast integer to string

# Display results
print(n)  
print(f)  
print(s2)
```

### Getting the Type of Variable

Dalam Python, kita dapat menentukan tipe suatu variabel menggunakan fungsi type(). Fungsi bawaan ini mengembalikan tipe objek yang diteruskan kepadanya.

Example Usage of type()

```python
# Define variables with different data types
n = 42
f = 3.14
s = "Hello, World!"
li = [1, 2, 3]
d = {'key': 'value'}
bool = True

# Get and print the type of each variable
print(type(n))   
print(type(f)) 
print(type(s))   
print(type(li))     
print(type(d))     
print(type(bool))
```

### Object Reference in Python

Let us assign a variable x to value 5.

x = 5

Saat x = 5dijalankan, Python membuat objek untuk mewakili nilai 5dan membuat xreferensi objek ini.

Sekarang, jika kita tetapkan variabel lain y ke variabel x.

y = x

Penjelasan:
+ Ketika Python menemukan pernyataan pertama, ia membuat objek untuk nilai tersebut 5dan xmereferensikannya. Pernyataan kedua membuat ydan mereferensikan objek yang sama dengan x, bukan xdirinya sendiri. Ini disebut Referensi Bersama , di mana beberapa variabel mereferensikan objek yang sama.

Sekarang, jika kita menulis

x = 'Orang aneh'

Python membuat objek baru untuk nilai tersebut "Geeks"dan membuat xreferensi ke objek baru ini.

Penjelasan:
+ Variabelnya ytetap tidak berubah, masih merujuk ke objek asli 5.
    
Jika sekarang kita tetapkan nilai baru ke y:

y = "Komputer"

+ Python membuat objek lain "Computer"dan memperbaruinya yuntuk mereferensikannya.
+ Objek asli 5tidak lagi memiliki referensi apa pun dan menjadi memenuhi syarat untuk pengumpulan sampah.

Poin-poin Utama:
+ Variabel Python menyimpan referensi ke objek, bukan objek itu sendiri.
+ Penugasan kembali suatu variabel tidak akan memengaruhi variabel lain yang merujuk ke objek yang sama kecuali diperbarui secara eksplisit.

### Delete a Variable Using del Keyword

Kita dapat menghapus variabel dari namespace menggunakan kata kunci del . Ini secara efektif menghapus variabel dan mengosongkan memori yang digunakannya.

Contoh:

```python
# Assigning value to variable
x = 10
print(x) 

# Removing the variable using del
del x

# Trying to print x after deletion will raise an error
# print(x)  # Uncommenting this line will raise NameError: name 'x' is not defined
```

Penjelasan:
+ del x menghapus variabel x dari memori.
+ Setelah dihapus, mencoba mengakses variabel x menghasilkan NameError, yang menunjukkan variabel tersebut tidak ada lagi.

### Practical Examples

#### 1. Swapping Two Variables

Dengan menggunakan beberapa penugasan, kita dapat menukar nilai dua variabel tanpa memerlukan variabel sementara.

```python
a, b = 5, 10
a, b = b, a
print(a, b)
```

#### 2. Counting Characters in a String

Tetapkan hasil beberapa operasi pada string ke variabel dalam satu baris.

```python
word = "Python"
length = len(word)
print("Length of the word:", length)
```

### Scope of a Variable

Dalam Python, cakupan suatu variabel menentukan di mana variabel tersebut dapat diakses dalam program. Ada dua jenis cakupan utama: lokal dan global .

#### Local Variables:

+ Didefinisikan dalam suatu fungsi atau blok, hanya dapat diakses dalam lingkup tersebut.
+ Dihancurkan setelah fungsi/blok berakhir.
+ Sementara, digunakan untuk data jangka pendek.

#### Global Variables:

+ Fungsi luar yang ditentukan, dapat diakses di seluruh program.
+ Untuk memodifikasi dalam suatu fungsi, gunakan globalkata kunci .
+ Bertahan dalam memori selama program berlangsung, berguna untuk data yang dibagikan.
