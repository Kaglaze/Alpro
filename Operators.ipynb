## 1.4 Python Operators

Dalam pemrograman Python, Operator umumnya digunakan untuk melakukan operasi pada nilai dan variabel. Operator adalah simbol standar yang digunakan untuk operasi logika dan aritmatika. Dalam artikel ini, kita akan membahas berbagai jenis operator Python. 
+ OPERATOR: Ini adalah simbol-simbol khusus. Misalnya +, *, /, dll.
+ OPERAND: Nilai tempat operator diterapkan.

### Types of Operators in Python

#### 1.Arithmetic Operators
Operator Aritmatika Python digunakan untuk melakukan operasi matematika dasar seperti penjumlahan, pengurangan, perkalian dan pembagian.

Pada Python 3.x, hasil pembagian adalah floating-point, sedangkan pada Python 2.x, pembagian 2 bilangan bulat adalah bilangan bulat. Untuk mendapatkan hasil bilangan bulat, pada Python 3.x, digunakan tanda lantai (// integer).

Contoh

```python
# Variables
a = 15
b = 4

# Addition
print("Addition:", a + b)  

# Subtraction
print("Subtraction:", a - b) 

# Multiplication
print("Multiplication:", a * b)  

# Division
print("Division:", a / b) 

# Floor Division
print("Floor Division:", a // b)  

# Modulus
print("Modulus:", a % b) 

# Exponentiation
print("Exponentiation:", a ** b)
```

Catatan: Lihat Perbedaan antara / dan // untuk beberapa fakta menarik tentang kedua operator Python ini.

#### 2.Comparison Operators
Dalam Python, operator perbandingan relasional membandingkan nilai. Operator ini menghasilkan nilai Benar atau Salah sesuai kondisi.

Contoh
Mari kita lihat contoh Operator Perbandingan dalam Python.

```python
a = 13
b = 33

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

#### 3.Logical Operators
Operator Logika Python menjalankan operasi Logika AND , Logika OR, dan Logika NOT . Operator ini digunakan untuk menggabungkan pernyataan kondisional.

Urutan Operator Logika dalam Python adalah sebagai berikut:

1. Logis tidak
2. logis dan
3. logis atau

```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

#### 4.Bitwise Operators
Operator Bitwise Python bekerja pada bit dan melakukan operasi bit demi bit. Operator ini digunakan untuk mengoperasikan bilangan biner.

Operator Bitwise dalam Python adalah sebagai berikut:
1. Bitwise TIDAK
2. Pergeseran Bitwise
3. Bitwise DAN
4. Bitwise XOR
5. Bitwise ATAU

```python
a = 10
b = 4

print(a & b)
print(a | b)
print(~a)
print(a ^ b)
print(a >> 2)
print(a << 2)
```

#### 5.Assignment Operators
Operator Penugasan Python digunakan untuk menetapkan nilai ke variabel. Operator ini digunakan untuk menetapkan nilai di sisi kanan ekspresi ke operan di sisi kiri.

Contoh

```python
a = 10
b = a
print(b)
b += a
print(b)
b -= a
print(b)
b *= a
print(b)
b <<= a
print(b)
```

#### 6.Identity Operators
Dalam Python, is dan is not adalah operator identitas. Keduanya digunakan untuk memeriksa apakah dua nilai terletak di bagian memori yang sama. Dua variabel yang sama tidak berarti keduanya identik. 

+ Benar           jika operannya identik, 
+ tidak       Benar jika operannya tidak identik

```python
a = 10
b = 20
c = a

print(a is not b)
print(a is c)
```

#### 7.Membership Operators
Dalam Python, in dan not in adalah operator keanggotaan yang digunakan untuk menguji apakah suatu nilai atau variabel ada dalam suatu urutan.

+ dalam             Benar jika nilai ditemukan dalam urutan 
+ tidak dalam         Benar jika nilai tidak ditemukan dalam urutan

```python
x = 24
y = 20
list = [10, 20, 30, 40, 50]

if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```

#### 8.Ternary Operator
Dalam Python, operator ternary, juga dikenal sebagai ekspresi kondisional, adalah operator yang mengevaluasi sesuatu berdasarkan kondisi yang benar atau salah. Operator ini ditambahkan ke Python pada versi 2.5. 

Ini hanya memungkinkan pengujian kondisi dalam satu baris menggantikan if-else multiline yang membuat kode menjadi ringkas.

Sintaksis:   [on_true] jika [ekspresi] jika tidak [on_false] 

```python
a, b = 10, 20
min = a if a < b else b

print(min)
```

## Precedence and Associativity of Operators
Dalam Python, prioritas dan asosiativitas Operator menentukan prioritas operator.

### Operator Precedence
Ini digunakan dalam ekspresi dengan lebih dari satu operator dengan prioritas berbeda untuk menentukan operasi mana yang akan dilakukan pertama.

```python
expr = 10 + 20 * 30
print(expr)
name = "Alex"
age = 0

if name == "Alex" or name == "John" and age >= 2:
    print("Hello! Welcome.")
else:
    print("Good Bye!!")
```

### Operator Associativity
Jika suatu ekspresi mengandung dua atau lebih operator dengan prioritas yang sama, maka Asosiativitas Operator digunakan untuk menentukannya. Operator dapat berupa Kiri ke Kanan atau Kanan ke Kiri.

```python
print(100 / 10 * 10)
print(5 - 2 + 3)
print(5 - (2 + 3))
print(2 ** 3 ** 2)
```
