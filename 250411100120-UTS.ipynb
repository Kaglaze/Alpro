```python
import random

print("--- Permainan Gunting Batu Kertas ---")
print("pilihan: gunting/batu/kertas")

nyawa_user = 3
nyawa_komputer = 3
ronde = 0

while True:
    user = input("pilih (gunting, batu, kertas):").lower()

    if user not in ["gunting", "batu", "kertas"]:
        print("Pilihan tidak valid, coba lagi!")
        continue

    komputer = random.choice(["gunting", "batu", "kertas"])
    print(f"Komputer Memilih: {komputer}")

    if user==komputer:
        print("seri!\n")
        ronde += 1
    elif(user == "gunting" and komputer == "kertas") or \
        (user == "kertas" and komputer =="batu") or \
        (user =="batu" and komputer == "gunting"):
        print("Kamu Menang ronde ini!\n")
        nyawa_komputer -= 1
        ronde += 1
    else:
        print("Komputer Menang ronde ini!\n")
        nyawa_user -= 1
        ronde += 1
        
        if nyawa_user != 0 or nyawa_komputer != 0:
            print(ronde)
            print("User Menang!")
        else:
            print("Komputer Menang!")

    print (f"Nyawa User: {nyawa_user}, Nyawa Komputer: {nyawa_komputer}\n")
```
