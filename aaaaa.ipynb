```python
import random

print("--- Permainan Gunting Batu Kertas ---")
print("pilihan: gunting/batu/kertas")
print("Ketik 'quit' untuk keluar dari permainan.\n")
#bikin ketentuan skor
skor_user = 0
skor_komputer = 0
batas_skor = 3

#pilihan permainan
while True:
    user = input("pilih (gunting, batu, kertas):").lower()
    
    if user == "quit":
        print("Permainan dihentikan.")
        break

    if user not in ["gunting", "batu", "kertas"]:
        print("Pilihan tidak valid, coba lagi!\n")
        continue

    komputer = random.choice(["gunting", "batu", "kertas"])
    print(f"Komputer Memilih: {komputer}")

    #alur permainan
    if user==komputer:
        print("seri!\n")
    elif(user == "gunting" and komputer == "kertas") or \
        (user == "kertas" and komputer =="batu") or \
        (user =="batu" and komputer == "gunting"):
        print("Kamu Menang!\n")
        skor_user += 1
    else:
        print("Komputer Menang!\n")
        skor_komputer += 1

#stop dan menampilkan skor
    print (f"Skor -> Kamu: {skor_user} | Komputer: {skor_komputer}\n")

    if skor_user == batas_skor:
        print("🎉 Selamat, kamu juara!")
        break
    elif skor_komputer == batas_skor:
        print("😅 Komputer juara, coba lagi ya!")
        break
```

```python
import random

angka_rahasia = random.randint(1, 100)

print("Tebak angka rahasia (antara 1 - 100)!")

while True:
    tebakan = int(input("Masukkan tebakanmu: "))
    
    if tebakan < angka_rahasia:
        print("Terlalu kecil!")
    elif tebakan > angka_rahasia:
        print("Terlalu besar!")
    else:
        print("Benar! Angka rahasianya adalah", angka_rahasia)
        break
```
