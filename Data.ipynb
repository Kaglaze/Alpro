###  Numeric Data Types

Angka dalam Python merepresentasikan data yang memiliki nilai numerik. Nilai numerik dapat berupa bilangan bulat, bilangan mengambang, atau bahkan bilangan kompleks. Nilai-nilai ini didefinisikan sebagai kelas int, float, dan kompleks.
+ Bilangan bulat - nilai direpresentasikan oleh kelas int. Nilai ini berisi bilangan bulat positif atau negatif (tanpa pecahan atau desimal). Tidak ada batasan panjang suatu nilai bilangan bulat.
+ Float - nilai direpresentasikan oleh kelas float. Float adalah bilangan riil dengan representasi floating-point. Nilai ini ditentukan oleh titik desimal. Secara opsional, karakter e atau E diikuti oleh bilangan bulat positif atau negatif dapat ditambahkan untuk menentukan notasi ilmiah.
+ Bilangan Kompleks - Direpresentasikan oleh kelas kompleks. Bilangan ini dispesifikasikan sebagai (bagian riil) + (bagian imajiner)j. Contoh: - 2+3j

```python
a = 5
print(type(a))

b = 5.0
print(type(b))

c = 2 + 4j
print(type(c))
```

### Sequence Data Types

Tipe Data Sequence adalah kumpulan berurutan dari tipe data Python yang serupa atau berbeda. Sequence memungkinkan penyimpanan beberapa nilai secara terorganisir dan efisien. Ada beberapa tipe data sequence dalam Python:

#### String Data Type

String Python adalah array byte yang mewakili karakter Unicode. Dalam Python, tidak ada tipe data karakter; karakter adalah string dengan panjang satu. String direpresentasikan oleh kelas str.

String dalam Python dapat dibuat menggunakan tanda kutip tunggal, tanda kutip ganda, atau bahkan tanda kutip tiga. Kita dapat mengakses karakter individual dari sebuah String menggunakan indeks.

```python
s = 'Welcome to the Geeks World'
print(s)

# check data type 
print(type(s))

# access string with index
print(s[1])
print(s[2])
print(s[-1])
```

#### List Data Type

List, seperti halnya array, dideklarasikan dalam bahasa pemrograman lain dan merupakan kumpulan data yang terurut. List sangat fleksibel karena item dalam list tidak harus bertipe sama.

##### Creating a List in Python

Daftar dalam Python dapat dibuat hanya dengan menempatkan urutan di dalam tanda kurung siku[].

```python
# Empty list
a = []

# list with int values
a = [1, 2, 3]
print(a)

# list with mixed int and string
b = ["Geeks", "For", "Geeks", 4, 5]
print(b)
```

##### Access List Items

Untuk mengakses item dalam daftar, rujuk nomor indeks. Dalam Python, indeks urutan negatif merepresentasikan posisi dari akhir array. Alih-alih harus menghitung offset seperti pada List[len(List)-3], cukup tulis List[-3]. Pengindeksan negatif berarti dimulai dari akhir, -1 merujuk ke item terakhir, -2 merujuk ke item kedua terakhir, dst.

```python
a = ["Geeks", "For", "Geeks"]
print("Accessing element from the list")
print(a[0])
print(a[2])

print("Accessing element using negative indexing")
print(a[-1])
print(a[-3])
```

#### Tuple Data Type

Tuple adalah kumpulan objek Python yang berurutan. Satu-satunya perbedaan antara tuple dan list adalah bahwa tuple tidak dapat diubah. Tuple tidak dapat diubah setelah dibuat.

##### Creating a Tuple in Python

Dalam Python, tupel dibuat dengan menempatkan serangkaian nilai yang dipisahkan oleh tanda koma, dengan atau tanpa tanda kurung untuk mengelompokkan urutan data. Tupel dapat berisi sejumlah elemen dan tipe data apa pun (seperti string, integer, list, dll.).

Catatan: Tuple juga dapat dibuat dengan satu elemen, tetapi agak rumit. Hanya memiliki satu elemen dalam tanda kurung saja tidak cukup, harus ada tanda koma di belakangnya agar menjadi tuple.

```python
# initiate empty tuple
tup1 = ()

tup2 = ('Geeks', 'For')
print("\nTuple with the use of String: ", tup2)
```

Catatan - Pembuatan tupel Python tanpa menggunakan tanda kurung dikenal sebagai Pengepakan Tupel.

##### Access Tuple Items

Untuk mengakses item tupel, rujuk nomor indeks. Gunakan operator indeks [ ] untuk mengakses item dalam tupel.

```python
tup1 = tuple([1, 2, 3, 4, 5])

# access tuple items
print(tup1[0])
print(tup1[-1])
print(tup1[-3])
```

### Set Data Type in Python

Dalam Tipe Data Python, Set adalah kumpulan tipe data yang tidak berurutan, dapat diulang, dapat diubah, dan tidak memiliki elemen duplikat. Urutan elemen dalam suatu set tidak ditentukan meskipun dapat terdiri dari berbagai elemen.

#### Create a Set in Python

Himpunan dapat dibuat menggunakan fungsi set() bawaan dengan objek yang dapat diiterasi atau suatu deret dengan menempatkan deret tersebut di dalam kurung kurawal, dipisahkan dengan tanda koma. Tipe elemen dalam suatu himpunan tidak harus sama; berbagai nilai tipe data yang tercampur juga dapat diteruskan ke dalam himpunan tersebut.

Himpunan dapat dibuat menggunakan fungsi set() bawaan dengan objek yang dapat diiterasi atau suatu deret dengan menempatkan deret tersebut di dalam kurung kurawal, dipisahkan dengan tanda koma. Tipe elemen dalam suatu himpunan tidak harus sama; berbagai nilai tipe data yang tercampur juga dapat diteruskan ke dalam himpunan tersebut.

Contoh: Kode ini adalah contoh cara membuat set menggunakan berbagai jenis nilai, seperti string, daftar, dan nilai campuran.

```python
# initializing empty set
s1 = set()

s1 = set("GeeksForGeeks")
print("Set with the use of String: ", s1)

s2 = set(["Geeks", "For", "Geeks"])
print("Set with the use of List: ", s2)
```

#### Access Set Items

Item-item dalam himpunan tidak dapat diakses dengan merujuk ke indeks, karena himpunan tidak berurutan, sehingga item-item tersebut tidak memiliki indeks. Namun, kita dapat melakukan perulangan melalui item-item dalam himpunan menggunakan for loop, atau menanyakan apakah suatu nilai tertentu ada dalam suatu himpunan, dengan menggunakan kata kunci in.

```python
set1 = set(["Geeks", "For", "Geeks"])
print(set1)

# loop through set
for i in set1:
    print(i, end=" ")
    
# check if item exist in set    
print("Geeks" in set1)
```

### Dictionary Data Type

Kamus dalam Python adalah kumpulan nilai data yang digunakan untuk menyimpan data seperti peta. Tidak seperti Tipe Data Python lainnya, Kamus menyimpan pasangan kunci:nilai. Kunci-nilai disediakan dalam Kamus agar lebih optimal. Setiap pasangan kunci-nilai dalam Kamus dipisahkan oleh titik dua:, sedangkan setiap kunci dipisahkan oleh koma.

#### Create a Dictionary in Python

Nilai dalam kamus dapat berupa tipe data apa pun dan dapat diduplikasi, sedangkan kunci tidak dapat diulang dan harus tetap. Kamus juga dapat dibuat dengan fungsi bawaan dict().

Catatan - Kunci kamus peka huruf besar-kecil, nama yang sama tetapi huruf besar-kecil Kunci yang berbeda akan diperlakukan secara berbeda.

```python
# initialize empty dictionary
d = {}

d = {1: 'Geeks', 2: 'For', 3: 'Geeks'}
print(d)

# creating dictionary using dict() constructor
d1 = dict({1: 'Geeks', 2: 'For', 3: 'Geeks'})
print(d1)
```

#### Accessing Key-value in Dictionary

Untuk mengakses item kamus, lihat nama kuncinya. Kunci dapat digunakan di dalam tanda kurung siku. Menggunakan metode get(), kita dapat mengakses elemen kamus.

```python
d = {1: 'Geeks', 'name': 'For', 3: 'Geeks'}

# Accessing an element using key
print(d['name'])

# Accessing a element using get
print(d.get(3))
```
