### 1.List
kata = ["data", "sains", "python", "analisis", "AI", "machine", "learning"]

Buat program untuk:
- Memisahkan kata yang memiliki panjang lebih dari 5 huruf dan kurang atau sama dengan 5 huruf
- Mengurutkan kedua kelompok berdasarkan panjang huruf
- Menggabungkan dua list tersebut sehingga kata terpanjang muncul terakhir

```python
kata = ["data", "sains", "python", "analisis", "AI", "machine", "learning"]

# 1
ksd5 = []
pld5 = []

for i in kata:
    if len(i) > 5:
        pld5.append(i)
    else:
        ksd5.append(i)
print(f"Kurang atau sama dengan 5: {ksd5}")
print(f"Panjang lebih dari 5: {pld5}")
print()

# 2
ksd5.sort()
pld5.sort(reverse = True)
print(f"Urutkan: {ksd5}")
print(f"Urutkan: {pld5}")
print()

# 3
gl = ksd5 + pld5
print(f"Gabungan List: {gl}")
```

### 3.List Comprehension
data = [" 24", "18 ", " python ", "99", " AI", " 7"]

Buat program dengan list comprehension untuk:
- Membersihkan spasi
- Memisahkan data menjadi dua kategori: numerik dan teks
- Mengubah data numerik menjadi bentuk integer dan teks menjadi huruf kapital

```python
data = [" 24", "18 ", " python ", "99", " AI", " 7"]

# 1
tspasi = [x.strip() for x in data]
print(f"Tanpa spasi: {tspasi}")
print()

# 2
numerik = [x if x.isdigit() else " " for x in tspasi]
teks = [x if x.isalpha() else "" for x in tspasi]
print(f"Numerik: {numerik}")
print(f"Teks: {teks}")
print()

# 3
integer = [x for x in numerik]
kapital = [x.upper() for x in teks]
print(f"Integer: {integer}")
print(f"Kapital: {kapital}")
```
