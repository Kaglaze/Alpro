#### Indentation in Python

Dalam Python, indentasi digunakan untuk mendefinisikan blok kode. Indentasi memberi tahu interpreter Python bahwa sekelompok pernyataan termasuk dalam blok tertentu. Semua pernyataan dengan tingkat indentasi yang sama dianggap sebagai bagian dari blok yang sama. Indentasi dicapai dengan menggunakan spasi (spasi atau tab) di awal setiap baris.

Misalnya:

```python
if 10 > 5:
    print("This is true!")
    print("I am tab indentation")

print("I have no indentation")
```

+ Dua pernyataan cetak pertama diberi indentasi 4 spasi, jadi keduanya termasuk dalam blok if.
+ Pernyataan cetak ketiga tidak menjorok, jadi berada di luar blok if.

Jika kita Melewati Indentasi, Python akan memunculkan kesalahan.

```python
if 10>5:
print("GeeksforGeeks")
```

#### Indentasi dalam Pernyataan Kondisional

Baris print( 'GeeksforGeeks…') dan print( 'retype the URL.' ) adalah dua blok kode terpisah. Kedua blok kode dalam contoh pernyataan if kita diberi jarak empat spasi. Baris terakhir print( 'All set!') tidak diberi jarak, sehingga tidak termasuk dalam blok else.

```python
a =  20

if a >= 18:
    print('GeeksforGeeks...')
else:
    print('retype the URL.')
print('All set !')
```

#### Indentation in Loops

Untuk menunjukkan blok kode dalam Python, kita harus menjorokkan setiap baris blok dengan spasi yang sama . Dua baris kode dalam perulangan while diberi indentasi empat spasi. Ini diperlukan untuk menunjukkan blok kode mana suatu pernyataan berada.

```python
j = 1
  
while(j<= 5): 
     print(j) 
     j = j + 1
```

### Historical Background
Python memiliki sejarah yang menarik. Pada tahun 1982, Guido van Rossum , pencipta Python, mulai bekerja di CWI , sebuah lembaga penelitian nasional Belanda. Ia bergabung dengan tim yang merancang bahasa pemrograman baru, bernama ABC, untuk pengajaran dan pembuatan prototipe. Kesederhanaan ABC ideal untuk pemula, tetapi bahasa tersebut tidak memiliki fitur yang dibutuhkan untuk menulis program tingkat lanjut.

Beberapa tahun kemudian, van Rossum bergabung dengan tim lain di CWI yang sedang mengerjakan sistem operasi. Tim tersebut membutuhkan cara yang lebih mudah untuk menulis program guna memantau komputer dan menganalisis data. Bahasa pemrograman yang umum pada tahun 1980-an (dan masih) sulit digunakan untuk program semacam ini. Van Rossum membayangkan sebuah bahasa pemrograman baru yang memiliki sintaksis sederhana, seperti ABC, tetapi juga menyediakan fitur-fitur canggih yang dibutuhkan para profesional.

Awalnya, van Rossum mulai mempelajari bahasa baru ini sebagai hobi di waktu luangnya. Ia menamai bahasa tersebut Python karena ia adalah penggemar grup komedi Inggris Monty Python . Selama setahun berikutnya, ia dan rekan-rekannya berhasil menggunakan Python berkali-kali dalam pekerjaan nyata. Van Rossum akhirnya memutuskan untuk berbagi Python dengan komunitas pemrograman yang lebih luas secara daring. Ia membagikan seluruh kode sumber Python secara gratis sehingga siapa pun dapat menulis dan menjalankan program Python.

Rilis pertama Python, yang dikenal sebagai Versi 0.9.0, muncul pada tahun 1991, sekitar enam tahun setelah C++ dan empat tahun sebelum Java. Keputusan van Rossum untuk membuat bahasa tersebut sederhana namun canggih, cocok untuk tugas sehari-hari, dan tersedia secara bebas daring berkontribusi pada keberhasilan jangka panjang Python.

### The hash character
Comments adalah frasa singkat yang menjelaskan apa yang dilakukan kode. Contoh: Baris 1, 8, dan 10 dalam program berikut berisi komentar. Setiap komentar diawali dengan karakter hash ( #). Semua teks dari karakter hash hingga akhir baris diabaikan saat program dijalankan. Sebaliknya, karakter hash di dalam string diperlakukan sebagai teks biasa. Contoh: String tersebut "Item #1: "tidak berisi komentar.

Saat menulis Comments:
+ Karakter #harus diikuti oleh satu spasi. Contoh: # End of menulebih mudah dibaca daripada #End of menu.
+ Komentar harus menjelaskan tujuan kode, bukan hanya mengulang kode itu sendiri. Contoh: # Get the user's preferenceslebih deskriptif daripada # Input item1 and item2.

Membuat kesalahan adalah bagian alami dari pemrograman. Bahkan programmer berpengalaman pun bisa membuat kesalahan saat menulis kode. Ketika terjadi kesalahan, Python menampilkan pesan dengan informasi berikut:
