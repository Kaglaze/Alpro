1. Buat program yang menerima dua bilangan (a, b), lalu menampilkan hasilnya untuk operator berikut + − × ÷ % Jika b == 0, tampilkan pesan “Tidak dapat dibagi dengan nol.” dan “Modulo tidak terdefinisi karena pembagi nol.”

```python
a = float(input("Masukkan angka: "))
b = float(input("Masukkan angka: "))

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b

if b != 0:
    pembagian = a / b
else:
    pembagian = "Tidak dapat dibagi dengan nol"

if b != 0:
    modulus = a % b
else:
    modulus = "Modulus tidak terdefinisi karena pembagi nol"
    
print(f"Hasil Penjumlahan: {penjumlahan}")
print(f"Hasil Pengurangan: {pengurangan}")
print(f"Hasil Perkalian: {perkalian}")
print(f"Hasil Pembagian: {pembagian}")
print(f"Hasil Modulus: {modulus}")
```

3. Buatkanlah Deret faktorial dari bilangan n bilangan + hitung rata rata dari bilangan tersebut

```python
def faktorial(x):
    """
    Fungsi untuk menghitung faktorial dari x.
    Menggunakan rekursi sederhana (untuk n kecil, <20, agar tidak overflow).
    """
    if x == 0 or x == 1:
        return 1
    return x * faktorial(x - 1)

def deret_faktorial_dan_rata_rata(n):
    """
    Menghitung deret faktorial 1! + 2! + ... + n! dan rata-ratanya.
    """
    if n < 1:
        return "n harus bilangan bulat positif!"
    
    total_deret = 0
    suku = []  # Untuk menyimpan suku-suku jika diperlukan
    
    for i in range(1, n + 1):
        f = faktorial(i)
        total_deret += f
        suku.append(f)
    
    rata_rata = total_deret / n
    return total_deret, rata_rata, suku

# Contoh penggunaan
n = int(input("Masukkan nilai n: "))  # Input dari user
total, rata, suku_list = deret_faktorial_dan_rata_rata(n)

print(f"Deret faktorial untuk n={n}: {suku_list}")
print(f"Total deret: {total}")
print(f"Rata-rata: {rata:.2f}")
```

4. Tulis program yang mencetak angka 1..10 hentikan perulangan lebih awal saat nilai mencapai 7 menggunakan break, lalu tampilkan pesan “Berhenti di 7”.

```python
i = 1
while i <= 10:
    print(i)
    i += 1
    if i == 7:
        print("Berhenti di 7")
        break
```
