# 7.1 Fungsi Python
Fungsi adalah blok kode yang hanya berjalan ketika dipanggil.

Suatu fungsi dapat mengembalikan data sebagai hasil.

Suatu fungsi membantu menghindari pengulangan kode.

## Membuat Fungsi
Dalam Python, sebuah fungsi didefinisikan menggunakan kata def kunci `function`, diikuti oleh nama fungsi dan tanda kurung:

```python
#Contoh

def my_function():
  print("Hello from a function")

my_function()
```

Ini menciptakan fungsi bernama my_functionyang mencetak "Halo dari suatu fungsi" saat dipanggil.

Kode di dalam fungsi harus menjorok. Python menggunakan indentasi untuk mendefinisikan blok kode.

## Memanggil sebuah Fungsi
Untuk memanggil sebuah fungsi, tuliskan namanya diikuti tanda kurung:

```python
#Contoh
def my_function():
  print("Hello from a function")

my_function()
```

```python
#Anda dapat memanggil fungsi yang sama beberapa kali:

#Contoh
def my_function():
  print("Hello from a function")

my_function()
my_function()
my_function()
```

## Mengapa Menggunakan Fungsi?
Bayangkan Anda perlu mengkonversi suhu dari Fahrenheit ke Celsius beberapa kali dalam program Anda. Tanpa fungsi, Anda harus menulis kode perhitungan yang sama berulang kali:

```python
#Contoh
#Tanpa fungsi - kode berulang:

temp1 = 77
celsius1 = (temp1 - 32) * 5 / 9
print(celsius1)

temp2 = 95
celsius2 = (temp2 - 32) * 5 / 9
print(celsius2)

temp3 = 50
celsius3 = (temp3 - 32) * 5 / 9
print(celsius3)
```

```python
#Dengan fungsi, Anda menulis kode sekali dan menggunakannya kembali:

#Contoh
#Dengan fungsi - kode yang dapat digunakan kembali:

def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius(77))
print(fahrenheit_to_celsius(95))
print(fahrenheit_to_celsius(50))
```

## Nilai Kembalian
Fungsi dapat mengirim data kembali ke kode yang memanggilnya menggunakan returnpernyataan.

Ketika suatu fungsi mencapai suatu returnpernyataan, ia berhenti mengeksekusi dan mengirimkan kembali hasilnya:

```python
#Contoh
#Fungsi yang mengembalikan nilai:

def get_greeting():
  return "Hello from a function"

message = get_greeting()
print(message)
```

```python
#Anda dapat menggunakan nilai yang dikembalikan secara langsung:

#Contoh
#Menggunakan nilai kembalian secara langsung:

def get_greeting():
  return "Hello from a function"

print(get_greeting())
```

Jika suatu fungsi tidak memiliki returnpernyataan, maka fungsi tersebut akan mengembalikan nilai Nonesecara default.

## Argumen Fungsi
Argumen adalah nilai yang dile传递kan di dalam tanda kurung fungsi. Sebuah fungsi dapat memiliki sejumlah argumen yang dipisahkan oleh koma.

Sintaksis:

def function_name(parameters): 

    """Docstring""" 
    
    # isi fungsi 
    
    return expression

Kita akan membuat fungsi sederhana di Python untuk memeriksa apakah angka yang diberikan sebagai argumen ke fungsi tersebut adalah genap atau ganjil.

```python
def evenOdd(x):
    if (x % 2 == 0):
        return "Even"
    else:
        return "Odd"

print(evenOdd(16))
print(evenOdd(7))
```

## Jenis-jenis Argumen Fungsi
Python mendukung berbagai jenis argumen yang dapat dilewatkan pada saat pemanggilan fungsi. Di Python, kita memiliki jenis argumen fungsi berikut, mari kita jelajahi satu per satu.

### 1. Argumen Default
Argumen default adalah parameter yang mengasumsikan nilai default jika nilai tidak diberikan dalam panggilan fungsi untuk argumen tersebut.

```python
def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10)
```

### 2. Argumen Kata Kunci
Dalam argumen kata kunci, nilai diteruskan dengan secara eksplisit menentukan nama parameter, sehingga urutannya tidak menjadi masalah.

```python
def student(fname, lname):
    print(fname, lname)

student(fname='Geeks', lname='Practice')
student(lname='Practice', fname='Geeks')
```

### 3. Argumen Posisional
Dalam argumen posisional, nilai diberikan kepada parameter berdasarkan urutannya dalam panggilan fungsi.

```python
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Suraj", 27)

print("\nCase-2:")
nameAge(27, "Suraj")
```

## Pernyataan Pengembalian dalam Fungsi
Pernyataan return mengakhiri sebuah fungsi dan mengirimkan nilai kembali ke pemanggil. Pernyataan ini dapat mengembalikan tipe data apa pun, beberapa nilai (dikemas dalam sebuah tuple), atau None jika tidak ada nilai yang diberikan.

Sintaksis:

kembalikan [ekspresi]

Parameter: return mengakhiri fungsi, [expression] adalah nilai opsional yang akan dikembalikan (nilai defaultnya adalah None) .

```python
def square_value(num):
    """This function returns the square
    value of the entered number"""
    return num**2

print(square_value(2))
print(square_value(-4))
```
