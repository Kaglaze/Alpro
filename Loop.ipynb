## 3.1 Loop

### Loops in Python - For, While and Nested Loops
Perulangan dalam Python digunakan untuk mengulang tindakan secara efisien. Jenis utamanya adalah perulangan For (menghitung item) dan perulangan While (berdasarkan kondisi). Dalam artikel ini, kita akan membahas perulangan Python dan memahami cara kerjanya dengan bantuan contoh.

### For Loop
Perulangan for digunakan untuk mengulang suatu urutan seperti daftar, tupel, string, atau rentang. Perulangan ini memungkinkan eksekusi blok kode berulang kali, sekali untuk setiap item dalam urutan tersebut.

Sintaksis:

untuk iterator_var dalam urutan: 
    pernyataan

```python
#Contoh:
n = 4
for i in range(0, n):
    print(i)
```

Penjelasan: Kode ini mencetak angka dari 0 hingga 3 (inklusif) menggunakan perulangan for yang berulang pada rentang 0 hingga n-1 (di mana n = 4).

Contoh:

Iterasi List, Tuple, String dan Dictionary Menggunakan Perulangan For di Python

```python
li = ["geeks", "for", "geeks"]
for i in li:
    print(i)
    
tup = ("geeks", "for", "geeks")
for i in tup:
    print(i)
    
s = "Geeks"
for i in s:
    print(i)
    
d = dict({'x':123, 'y':354})
for i in d:
    print("%s  %d" % (i, d[i]))
    
set1 = {1, 2, 3, 4, 5, 6}
for i in set1:
    print(i),
```

#### Iterating by Index of Sequences
Kita juga dapat menggunakan indeks elemen dalam urutan tersebut untuk melakukan iterasi. Ide kuncinya adalah menghitung panjang daftar terlebih dahulu, lalu melakukan iterasi pada urutan dalam rentang panjang tersebut.

```python
li = ["geeks", "for", "geeks"]
for index in range(len(li)):
    print(li[index])
```

Penjelasan: Kode ini mengiterasi setiap elemen daftar menggunakan indeksnya dan mencetak setiap elemen satu per satu. Fungsi range(len(daftar)) menghasilkan indeks dari 0 hingga panjang daftar dikurangi 1.

#### Using else Statement with for Loop
Kita juga bisa menggabungkan pernyataan else dengan perulangan for seperti pada perulangan while. Namun, karena tidak ada kondisi dalam perulangan for yang menentukan akhir eksekusi, blok else akan dieksekusi segera setelah blok for selesai dieksekusi.

```python
li = ["geeks", "for", "geeks"]
for index in range(len(li)):
    print(li[index])
else:
    print("Inside Else Block")
```

Penjelasan: Kode mengulangi daftar dan mencetak setiap elemen. Setelah perulangan berakhir, kode akan mencetak "Di Dalam Blok Else" karena blok else dieksekusi ketika perulangan selesai tanpa jeda.

### While Loop
Dalam Python, perulangan while digunakan untuk mengeksekusi blok pernyataan berulang kali hingga kondisi yang ditentukan terpenuhi. Ketika kondisinya menjadi salah, baris setelah perulangan dalam program akan dieksekusi.

Sintaksis:

ekspresi while: 
    pernyataan

Semua pernyataan yang diberi indentasi dengan jumlah spasi karakter yang sama setelah konstruksi pemrograman dianggap sebagai bagian dari satu blok kode. Python menggunakan indentasi sebagai metode pengelompokan pernyataan. 

Contoh:
Contoh ini mendemonstrasikan perulangan while dasar dalam Python. Perulangan ini berjalan selama kondisi cnt < 3 bernilai true. Perulangan ini menambahkan 1 pada penghitung pada setiap iterasi dan menampilkan "Hello Geek" tiga kali.

```python
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Geek")
```

#### Using else statement with While Loop
Klausa else hanya dieksekusi ketika kondisi while kita bernilai salah. Jika kita keluar dari loop atau jika pengecualian dimunculkan, maka klausa else tidak akan dieksekusi. 

Sintaksis:

sementara kondisi: 

     # jalankan pernyataan ini 
yang lain: 

     # jalankan pernyataan ini

Contoh:

Kode tersebut mencetak "Hello Geek" tiga kali menggunakan perulangan 'while' dan kemudian setelah perulangan tersebut, kode tersebut mencetak "In Else Block" karena ada blok "else" yang dikaitkan dengan perulangan 'while'.

```python
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Geek")
else:
    print("In Else Block")
```

### Infinite While Loop
Jika kita menginginkan suatu blok kode dieksekusi tak terhingga kali, maka kita dapat menggunakan loop while di Python untuk melakukannya.

Kode yang diberikan di bawah ini menggunakan perulangan 'while' dengan kondisi " True ", yang berarti perulangan akan berjalan tanpa henti hingga kita keluar darinya menggunakan kata kunci " break " atau logika lainnya.

```python
while (True):
    print("Hello Geek")
```

Catatan: Disarankan untuk tidak menggunakan jenis perulangan ini karena merupakan perulangan tak terhingga yang tidak pernah berakhir di mana kondisinya selalu benar dan kita harus menghentikan kompilator secara paksa.

### Nested Loops
Bahasa pemrograman Python memungkinkan penggunaan satu loop di dalam loop lain, yang disebut loop bersarang . Bagian berikut menunjukkan beberapa contoh untuk mengilustrasikan konsep tersebut. 

Sintaksis:

untuk iterator_var secara berurutan: 
   untuk iterator_var secara berurutan: 
       pernyataan 
   pernyataan

Sintaks untuk pernyataan while loop bersarang dalam bahasa pemrograman Python adalah sebagai berikut: 

ekspresi while: 
   ekspresi while: 
       pernyataan 
   pernyataan

Catatan terakhir tentang perulangan bersarang adalah kita dapat menempatkan jenis perulangan apa pun di dalam jenis perulangan lain di Python. Misalnya, perulangan for dapat berada di dalam perulangan while atau sebaliknya.

```python
from __future__ import print_function
for i in range(1, 5):
    for j in range(i):
        print(i, end=' ')
    print()
```

Penjelasan: Dalam kode di atas, kita menggunakan loop bersarang untuk mencetak nilai i beberapa kali di setiap baris, dengan jumlah pencetakan i meningkat seiring dengan setiap iterasi loop luar. Fungsi print() mencetak nilai i dan berpindah ke baris berikutnya setelah setiap baris.

### Loop Control Statements
Pernyataan kontrol loop mengubah eksekusi dari urutan normalnya. Ketika eksekusi meninggalkan suatu ruang lingkup, semua objek otomatis yang dibuat dalam ruang lingkup tersebut akan dihancurkan. Python mendukung pernyataan kontrol berikut. 

#### Continue Statement
Pernyataan continue dalam Python mengembalikan kontrol ke awal loop.

Contoh:

```python
for letter in 'geeksforgeeks':
    if letter == 'e' or letter == 's':
        continue
    print('Current Letter :', letter)
```

Penjelasan: Pernyataan continue digunakan untuk melewati iterasi loop saat ini dan melanjutkan ke iterasi berikutnya. Pernyataan ini berguna ketika kita ingin melewati kondisi tertentu tanpa menghentikan loop.

#### Break Statement
Pernyataan break dalam Python membawa kontrol keluar dari loop.

Contoh:

```python
for letter in 'geeksforgeeks':
    if letter == 'e' or letter == 's':
        break

print('Current Letter :', letter)
```

Penjelasan: Pernyataan break digunakan untuk keluar dari loop sebelum waktunya ketika kondisi tertentu terpenuhi. Dalam contoh ini, loop berhenti ketika hurufnya adalah 'e' atau 's', sehingga menghentikan iterasi selanjutnya.

#### Pass Statement
Kita menggunakan pernyataan pass dalam Python untuk menulis perulangan kosong. Pass juga digunakan untuk pernyataan kontrol, fungsi, dan kelas yang kosong.

Contoh:

```python
for letter in 'geeksforgeeks':
    pass
print('Last Letter :', letter)
```

Penjelasan: Dalam contoh ini, perulangan mengulangi setiap huruf dalam 'geeksforgeeks' tetapi tidak melakukan operasi apa pun, dan setelah perulangan selesai, huruf terakhir ('s') dicetak.

#### How for loop works internally?
Sebelum melanjutkan ke bagian ini, kita perlu memahami terlebih dahulu tentang Iterator Python. Pertama, mari kita lihat seperti apa perulangan for sederhana dalam Python.

Contoh 1:

Kode Python ini menelusuri daftar buah-buahan, yang berisi "apel", "jeruk", dan "kiwi". Kode ini mencetak setiap nama buah pada baris terpisah, menampilkannya sesuai urutan kemunculannya dalam daftar.

```python
fruits = ["apple", "orange", "kiwi"]

for fruit in fruits:

    print(fruit)
```

Kode ini mengulangi setiap item dalam daftar buah-buahan dan mencetak item (buah) pada setiap iterasi dan output akan menampilkan setiap buah pada baris baru.

Contoh 2:

Kode Python ini melakukan iterasi secara manual melalui daftar buah menggunakan sebuah iterator. Kode ini mencetak nama setiap buah satu per satu dan berhenti ketika tidak ada lagi item dalam daftar.

```python
fruits = ["apple", "orange", "kiwi"]
iter_obj = iter(fruits)
while True:
    try:
        fruit = next(iter_obj)
        print(fruit)
    except StopIteration:
        break
```
