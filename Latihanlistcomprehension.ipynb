1. Buat list berisi kelipatan3 dari angka 1 sampai 10 menggunakan list comprehension.


```python
newlist = [x for x in range(1, 10) if x % 3 == 0]
print(newlist)
```

2. Dari list nums = [1, 2, 3, 4, 5], buat list baru berisi setiap elemen dikali 3.

```python
nums = [1, 2, 3, 4, 5]
newlist = [x * 3 for x in nums]
print(newlist)
```

3. Dari list items = ['apple', 'banana','kiwi'], buat list berisi panjang masing-masing string.

```python
items = ['apple', 'banana','kiwi']
newlist = [len(x) for x in items]
print(newlist)
```

4. Buat list baru dari angka 1-20 yang hanya berisi bilangan genap.

```python
newlist = [x for x in range(1, 21) if x % 2 == 0]
print(newlist)
```

5. Buat list berisi karakter huruf dari string 'python'.

```python
newlist = [x for x in 'python']
print(newlist)
```

6. Dari nums = [1,2,3,4,5,6,7], buat list berisi angka > 3 dan dikalikan 2.

```python
nums = [1,2,3,4,5,6,7]
newlist = [x > 3 and x * 2 for x in nums]
print(newlist)
```

7. Flatten mat = [[1,2], [3,4], [5,6]] menjadi list 1D.

```python
mat = [[1,2], [3,4], [5,6]]
newlist = [x for i in mat for x in i]
print(newlist)
```
