## 2.2 Python Data Types

### Learning objectives

Pada akhir bagian ini Anda seharusnya dapat
+ Mengidentifikasi tipe data yang dihasilkan oleh operasi dengan bilangan bulat, bilangan float, dan string.
+ Gunakan operator dan konversi tipe untuk menggabungkan bilangan bulat, bilangan float, dan string.

### Combining integers and floats

Programmer sering kali perlu menggabungkan angka-angka dengan tipe data yang berbeda. Contoh: Sebuah program menghitung total pesanan belanja daring:

    kuantitas = int ( input ())
    harga = float ( input ())
    total = jumlah * harga
    cetak(total)
quantityadalah bilangan bulat, dan pricemerupakan bilangan float. Jadi, apa tipe data dari total? Untuk input 3dan 5.0, totaladalah bilangan float, dan program mencetak 15.0.

Menggabungkan bilangan bulat dan bilangan float menghasilkan bilangan float. Bilangan float secara default dicetak dengan setidaknya satu angka setelah koma desimal dan memiliki angka sebanyak yang diperlukan untuk mewakili nilai tersebut. Catatan: Pembagian menggunakan /operator selalu menghasilkan bilangan float.

### Combining numeric types and strings

Konversi tipe yang mudah dalam Python dapat membuat programmer berasumsi bahwa setiap tipe data dapat digabungkan dengan tipe data lain. Contoh: Program Noor membaca angka dari input dan menggunakan angka tersebut dalam perhitungan. Hal ini mengakibatkan kesalahan dalam program karena fungsi input()secara default menyimpan angka tersebut sebagai string. Tipe data string dan numerik tidak kompatibel untuk penjumlahan, pengurangan, dan pembagian. Salah satu operan perlu dikonversi secara eksplisit tergantung pada tujuan aritmatika atau penggabungan string.

Operator * juga berfungsi sebagai operator pengulangan, yang menerima operan string dan operan integer, lalu mengulang string tersebut. Contoh: menghasilkan ."banjo" * 3"banjobanjobanjo"
