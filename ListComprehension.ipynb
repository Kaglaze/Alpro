## Pemahaman Daftar
Pemahaman daftar menawarkan sintaksis yang lebih pendek saat Anda ingin membuat daftar baru berdasarkan nilai daftar yang ada.

Contoh:

Berdasarkan daftar buah-buahan, Anda menginginkan daftar baru yang hanya berisi buah-buahan dengan huruf "a" dalam namanya.

Tanpa pemahaman daftar, Anda harus menulis forpernyataan dengan pengujian kondisional di dalam:

```python
#Contoh
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

Dengan pemahaman daftar, Anda dapat melakukan semua itu hanya dengan satu baris kode:

```python
#Contoh
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x for x in fruits if "a" in x]

print(newlist)
```

## Sintaksis
newlist = [expression for item in iterable if condition == True]

Nilai yang dikembalikan adalah daftar baru, yang tidak mengubah daftar lama.

## Kondisi
Kondisi ini seperti filter yang hanya menerima item yang bernilai True.

```python
#Contoh
#Hanya terima barang yang bukan "apple":

newlist = [x for x in fruits if x != "apple"]
print(newlist)
```

Kondisi if x != "apple"  akan kembali Trueuntuk semua elemen selain "apel", membuat daftar baru berisi semua buah kecuali "apel".

Kondisi ini bersifat opsional dan dapat dihilangkan:

```python
#Contoh
#Tanpa ifpernyataan:

newlist = [x for x in fruits]
print(newlist)
```

## Dapat diulang
Yang dapat diulang dapat berupa objek apa saja, seperti daftar, tupel, himpunan, dan sebagainya.

```python
#Contoh
#Anda dapat menggunakan range()fungsi ini untuk membuat sesuatu yang dapat diulang:

newlist = [x for x in range(10)]
print(newlist)
```

Contoh yang sama, tetapi dengan suatu kondisi:

```python
#Contoh
#Hanya terima angka di bawah 5:

newlist = [x for x in range(10) if x < 5]
print(newlist)
```

## Ekspresi
Ekspresi adalah item saat ini dalam iterasi, tetapi juga merupakan hasil, yang dapat Anda manipulasi sebelum berakhir seperti item daftar dalam daftar baru:

```python
#Contoh
#Tetapkan nilai dalam daftar baru ke huruf besar:

newlist = [x.upper() for x in fruits]
print(newlist)
```

Anda dapat mengatur hasilnya sesuai keinginan Anda:

```python
#Contoh
#Tetapkan semua nilai dalam daftar baru ke 'halo':

newlist = ['hello' for x in fruits]
print(newlist)
```

Ekspresi tersebut juga dapat berisi kondisi, tidak seperti filter, tetapi sebagai cara untuk memanipulasi hasilnya:

```python
#Contoh
#Kembalikan "orange" bukannya "banana":

newlist = [x if x != "banana" else "orange" for x in fruits]
print(newlist)
```

Ekspresi pada contoh di atas mengatakan :

"Kembalikan barangnya kalau bukan pisang, kalau pisang kembalikan jeruk".

## Perulangan Menggunakan Pemahaman Daftar
Pemahaman Daftar menawarkan sintaksis terpendek untuk melakukan pengulangan melalui daftar:

```python
#Contoh
#Loop tangan pendek foryang akan mencetak semua item dalam daftar:

thislist = ["apple", "banana", "cherry"]
[print(x) for x in thislist]
```

## Pemahaman Daftar dalam Python

Pemahaman daftar adalah cara yang ringkas dan ampuh untuk membuat daftar baru dengan menerapkan ekspresi ke setiap item dalam iterabel yang sudah ada (seperti daftar, tupel, atau rentang). Pemahaman daftar membantu Anda menulis kode yang bersih, mudah dibaca, dan efisien dibandingkan dengan perulangan tradisional.

Misalkan Anda ingin mengkuadratkan setiap angka dalam suatu daftar:

```python
a = [2,3,4,5]
res = [val ** 2 for val in a]
print(res)
```

Penjelasan:res = [val ** 2 untuk val dalam a] gunakan pemahaman daftar untuk membuat daftar baru dengan mengkuadratkan setiap angka dalam .

### Sintaksis
[ekspresi untuk item dalam kondisi yang dapat diulang jika]

### Parameternya:

- ekspresi: operasi atau nilai yang akan disertakan dalam daftar baru.
- item: elemen saat ini dari yang dapat diulang.
- dapat diulang: urutan seperti daftar, tupel atau rentang.
- jika kondisi (opsional): filter untuk hanya menyertakan item yang memenuhi kondisi.

### Mengapa Menggunakan Pemahaman Daftar?
- Kode yang lebih bersih: Menggabungkan perulangan, penyaringan, dan transformasi dalam satu baris.
- Lebih mudah dibaca: Menghindari pengulangan yang bertele-tele dan variabel sementara.
- Eksekusi lebih cepat: Seringkali lebih cepat daripada for-loop yang setara.

### Perulangan For vs. Pemahaman Daftar
Perulangan for membutuhkan beberapa baris untuk membuat daftar baru dengan mengulang dan menambahkan setiap item secara manual. Pemahaman daftar melakukan hal yang sama hanya dalam satu baris, sehingga kode menjadi lebih pendek dan mudah dibaca.

Berikut ini contoh untuk menggandakan setiap angka dalam daftar:

### Menggunakan perulangan For

```python
a = [1, 2, 3, 4, 5]
res = []
for val in a:
    res.append(val * 2)
print(res)
```

Penjelasan:
- res = [] membuat daftar kosong untuk menyimpan hasil.
- for val in a: mengulang setiap angka dalam daftar a .
- res.append(val * 2) menggandakan angka saat ini val dan menambahkannya ke daftar res .

### Menggunakan Pemahaman Daftar

```python
a = [1, 2, 3, 4, 5]
res = [val * 2 for val in a]
print(res)
```

Penjelasan: res = [val * 2 untuk val dalam a] gunakan pemahaman daftar untuk membuat daftar baru dengan menggandakan setiap angka dalam a

## Pernyataan Kondisional dalam Pemahaman Daftar
Pemahaman daftar dapat menggunakan kondisi untuk memilih atau mengubah item berdasarkan aturan tertentu. Hal ini memungkinkan pembuatan daftar kustom yang lebih ringkas dan meningkatkan keterbacaan serta efisiensi kode.

Kode ini menggunakan pemahaman daftar dengan kondisi untuk membuat daftar baru yang hanya berisi angka genap dari daftar asli.

```python
a = [1, 2, 3, 4, 5]
res = [val for val in a if val % 2 == 0]
print(res)
```

Penjelasan: [val for val in a if val % 2 == 0] hanya memilih angka genap dari daftar a .

## Contoh pemahaman daftar
### 1. Membuat daftar dari suatu rentang
Kita dapat dengan cepat membuat daftar angka dalam rentang tertentu menggunakan pemahaman daftar. Contoh ini menghasilkan angka dari 0 hingga 9 dan menyimpannya dalam sebuah daftar.

```python
a = [i for i in range(10)]
print(a)
```

### 2. Menggunakan loop bersarang
Daftar semua pasangan koordinat dalam kisi 3x3 dapat dihasilkan dengan menggabungkan dua loop di dalam pemahaman daftar. Contoh ini menghasilkan setiap pasangan (x, y) yang memungkinkan, di mana x dan y berada dalam rentang 0 hingga 2.

```python
c = [(x, y) for x in range(3) for y in range(3)]
print(c)
```

### 3. Meratakan daftar daftar
Daftar bersarang (matriks) dapat diubah menjadi satu daftar datar dengan melakukan iterasi pada setiap subdaftar dan elemen-elemennya. Contoh ini meratakan matriks 3x3 menjadi satu daftar nilai yang berkesinambungan.

```python
mat = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
res = [val for row in mat for val in row]
print(res)
```

Penjelasan:

- mat adalah daftar dari daftar (matriks 3x3).
- [val untuk baris dalam mat untuk val dalam baris] menelusuri setiap subdaftar (baris) dan kemudian setiap nilai (val) di dalamnya.
- Mengumpulkan semua nilai ke dalam satu daftar datar.
