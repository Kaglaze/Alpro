Buatlah fungsi untuk menghitung faktorial suatu bilangan seperti contoh berikut 

5! = 5 x 4 x 3 x 2 x 1 = 120

3! = 3 x 2 x 1 = 6

2! = 2 x 1 = 2

0! = 1

```python
def factorial(num):
    if num==0:
        temp=1
    else:
        temp=1
        while num>=1:
            temp=temp*num
            num=num-1
    return(temp)
# Main Program #
a=int(input("Masukkan bilangan="))
b=factorial(a)
print (a,'!=',b)
```

+ Buat fungsi iteratif dan rekursif untuk menghitung eksponential suatu bilangan (x**n). Fungsi yang dibuat memiliki dua buah argument, yaitu x dan n

```python
def eksponensial_iteratif(x, n):
    hasil = 1
    for i in range(n):
        hasil *= x
    return hasil
    
x = int(input("Masukkan x: "))
n = int(input("Masukkan n: "))
total = eksponensial_iteratif(x, n)
print(f"x**n = {total}")
```

```python
def eksponensial_rekursif(x, n):
    if n == 0:
        return 1
    else:
        return x * eksponensial_rekursif(x, n - 1)
        
x = int(input("Masukkan x: "))
n = int(input("Masukkan n: "))
total = eksponensial_rekursif(x, n)
print(f"x**n = {total}")
```

```python
def eksponensial_rekursif(x, n):
    if n == 0:
        return 1
    else:
        return x * eksponensial_rekursif(x, n - 1)

def proses_perhitungan(x, n):
    if n == 0:
        return "1" 
    return " * ".join([str(x)] * n)

x = int(input("Masukkan x: "))
n = int(input("Masukkan n: "))

total = eksponensial_rekursif(x, n)
proses = proses_perhitungan(x, n)

print(f"{x}**{n} = {proses} = {total}")
```

```python
def jumlah_deret(n):
    if n == 0:
        return 0
    else:
        return n + jumlah_deret(n - 1)

def proses_jumlah_deret(n):
    return " + ".join(str(i) for i in range(1, n + 1))

n = int(input("Masukkan n: "))
total = jumlah_deret(n)
proses = proses_jumlah_deret(n)

print(f"Jumlah 1 sampai {n} = {proses} = {total}")
```

```python
def kali_rekursif(a, b):
    if b == 0:
        return 0
    else:
        return a + kali_rekursif(a, b - 1)

def proses_kali(a, b):
    return " + ".join([str(a)] * b)

a = int(input("Masukkan a: "))
b = int(input("Masukkan b: "))

total = kali_rekursif(a, b)
proses = proses_kali(a, b)

print(f"{a} * {b} = {proses} = {total}")
```
