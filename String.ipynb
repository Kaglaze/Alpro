## 4.1 String

Dalam Python, string adalah serangkaian karakter yang diapit tanda kutip. String dapat berisi huruf, angka, simbol, atau spasi. Karena Python tidak memiliki tipe karakter terpisah, bahkan satu karakter pun dianggap sebagai string dengan panjang satu. String banyak digunakan untuk penanganan dan manipulasi teks.

### Membuat String
String dapat dibuat menggunakan tanda kutip tunggal ('...') atau ganda ("...") . Keduanya berfungsi sama.

Contoh : Membuat dua string yang ekuivalen, satu dengan tanda kutip tunggal dan lainnya dengan tanda kutip ganda.

```python
s1 = 'GfG'  # single quote
s2 = "GfG"  # double quote
print(s1)
print(s2)
```

### Multi-line Strings
Gunakan tanda kutip tiga ('''...''') atau ( """...""") untuk string yang mencakup beberapa baris. Baris baru dipertahankan.

Contoh: Menentukan dan mencetak string multi-baris menggunakan kedua gaya.

```python
s = """I am Learning
Python String on GeeksforGeeks"""
print(s)

s = '''I'm a 
Geek'''
print(s)
```

### Accessing characters in String
String adalah urutan yang terindeks. Indeks positif dimulai dari 0 dari kiri ; indeks negatif dimulai dari -1 dari kanan

Contoh 1: Akses karakter tertentu melalui pengindeksan positif.

```python
s = "GeeksforGeeks"
print(s[0])   # first character
print(s[4])   # 5th character
```

Catatan: Mengakses indeks di luar rentang akan menyebabkan IndexError. Hanya bilangan bulat yang diperbolehkan sebagai indeks, dan penggunaan float atau tipe data lainnya akan mengakibatkan TypeError.

Contoh 2: Membaca karakter dari akhir menggunakan indeks negatif .

```python
s = "GeeksforGeeks"
print(s[-10])   # 3rd character
print(s[-5])    # 5th character from end
```

### String Slicing
Slicing adalah cara untuk mengekstrak sebagian string dengan menentukan indeks awal dan akhir . Sintaks untuk pengirisan adalah string[mulai:akhir] , di mana mulai adalah indeks awal dan akhir adalah indeks akhir (dikecualikan).

Contoh: Dalam contoh ini kita mengiris rentang dan membalikkan string.

```python
s = "GeeksforGeeks"
print(s[1:4])    # characters from index 1 to 3
print(s[:3])     # from start to index 2
print(s[3:])     # from index 3 to end
print(s[::-1])   # reverse string
```

### Iterasi String
String dapat diulang; Anda dapat mengulang karakter satu demi satu.

Contoh: Di sini, ia mencetak setiap karakter pada barisnya sendiri.

```python
s = "Python"
for char in s:
    print(char)
```

Penjelasan: perulangan for menarik karakter secara berurutan dan setiap iterasi mencetak karakter berikutnya.

### String Immutability
String bersifat immutable, artinya string tidak dapat diubah setelah dibuat. Jika kita perlu memanipulasi string, kita dapat menggunakan metode seperti penggabungan, pemotongan , atau pemformatan untuk membuat string baru berdasarkan string asli.

Contoh: Dalam contoh ini kita mengubah karakter pertama dengan membuat string baru.

```python
s = "geeksforGeeks"
s = "G" + s[1:]   # create new string
print(s)
```

### Menghapus String
Dalam Python, menghapus karakter individual dari sebuah string tidak mungkin dilakukan karena string bersifat immutable. Namun, kita dapat menghapus seluruh variabel string menggunakan kata kunci del .

Contoh: Di sini, kita menggunakan kata kunci del untuk menghapus string.

```python
s = "GfG"
del s
```

Catatan: Setelah menghapus string, jika kita mencoba mengakses s maka akan mengakibatkan NameError karena variabel tidak ada lagi.

### Memperbarui String
Karena string tidak dapat diubah, “pembaruan” membuat string baru menggunakan pemotongan atau metode seperti replace() .

Contoh: Kode ini memperbaiki huruf pertama dan mengganti kata.

```python
s = "hello geeks"
s1 = "H" + s[1:]                   # update first character
s2 = s.replace("geeks", "GeeksforGeeks")  # replace word
print(s1)
print(s2)
```

Penjelasan:
+ s1: potong dari indeks 1 dan seterusnya dan tambahkan "H".
+ s2: replace("geeks", "GeeksforGeeks") mengembalikan string baru.

### Metode String Umum
Python menyediakan berbagai metode bawaan untuk memanipulasi string. Berikut adalah beberapa metode yang paling berguna:

1. len() : Fungsi len() mengembalikan jumlah total karakter dalam sebuah string (termasuk spasi dan tanda baca).

```python
#Contoh
s = "GeeksforGeeks"
print(len(s))
```

2. upper() dan lower() : metode upper() mengubah semua karakter menjadi huruf besar sedangkan metode lower() mengubah semua karakter menjadi huruf kecil.

```python
s = "Hello World"
print(s.upper())
print(s.lower())
```

3. strip() dan replace() : strip() menghapus spasi di awal dan akhir string dan replace() mengganti semua kemunculan sub-string yang ditentukan dengan yang lain.

```python
s = "   Gfg   "
print(s.strip())    

s = "Python is fun"
print(s.replace("fun", "awesome"))
```

### Menggabungkan dan Mengulangi String
Kita dapat menggabungkan string menggunakan operator + dan mengulanginya menggunakan operator *.
1. String dapat digabungkan dengan menggunakan operator + .

Contoh: Gabungkan dua kata dengan spasi.

```python
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)
```

2. Kita dapat mengulang string beberapa kali menggunakan operator * .

Contoh: Mengulangi salam sebanyak tiga kali.

```python
s = "Hello "
print(s * 3)
```

### Memformat String
Python menyediakan beberapa cara untuk memasukkan variabel ke dalam string.
1. Menggunakan senar f

Cara paling sederhana dan paling disukai untuk memformat string adalah dengan menggunakan f-string .

Contoh: Sematkan variabel secara langsung menggunakan placeholder {}.

```python
name = "Alice"
age = 22
print(f"Name: {name}, Age: {age}")
```

2. Menggunakan format()
Cara lain untuk memformat string adalah dengan menggunakan metode format()

Contoh: Gunakan placeholder {} dan berikan nilai secara posisional.

```python
s = "My name is {} and I am {} years old.".format("Alice", 22)
print(s)
```

### Pengujian Keanggotaan String
Kata kunci in memeriksa apakah substring tertentu hadir dalam sebuah string.

Contoh: Di sini, kami menguji keberadaan substring.

```python
s = "GeeksforGeeks"
print("Geeks" in s)
print("GfG" in s)
```
