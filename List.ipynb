# 5.1 Python Lists

## Lists
Lists digunakan untuk menyimpan beberapa item dalam satu variabel.

Lists adalah salah satu dari 4 tipe data bawaan dalam Python yang digunakan untuk menyimpan kumpulan data, 3 lainnya adalah Tuple , Set , dan Dictionary , semuanya dengan kualitas dan penggunaan yang berbeda.

Lists dibuat menggunakan tanda kurung siku:

```python
#Contoh
#Buat Daftar:

thislist = ["apple", "banana", "cherry"]
print(thislist)
```

## Daftar Item
Item dalam daftar diurutkan, dapat diubah, dan memperbolehkan nilai duplikat.

Item daftar diindeks, item pertama memiliki indeks [0], item kedua memiliki indeks, [1]dan seterusnya.

## Dipesan
Ketika kita katakan bahwa daftar itu diurutkan, artinya item-itemnya mempunyai urutan yang pasti, dan urutan itu tidak akan berubah.

Jika Anda menambahkan item baru ke suatu daftar, item baru tersebut akan ditempatkan di akhir daftar.

Catatan: Ada beberapa metode daftar yang akan mengubah urutan, tetapi secara umum: urutan item tidak akan berubah.

## Dapat diubah
Daftar tersebut dapat diubah, artinya kita dapat mengubah, menambah, dan menghapus item dalam daftar setelah daftar dibuat.

## Izinkan Duplikat
Karena daftar diindeks, daftar dapat memiliki item dengan nilai yang sama:

```python
#Contoh
#Daftar memperbolehkan nilai duplikat:

thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```

## Panjang Daftar
Untuk menentukan berapa banyak item yang dimiliki suatu daftar, gunakan len()fungsi:

```python
#Contoh
#Cetak jumlah item dalam daftar:

thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

## Item Daftar - Tipe Data
Item daftar dapat berupa tipe data apa pun:

```python
#Contoh
#Tipe data string, int dan boolean:

list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
```

Suatu daftar dapat berisi berbagai tipe data:

```python
#Contoh
#Daftar dengan string, integer, dan nilai boolean:

list1 = ["abc", 34, True, 40, "male"]
```

## jenis()
Dari perspektif Python, daftar didefinisikan sebagai objek dengan tipe data 'daftar':

<class 'list'>

```python
#Contoh
#Apa tipe data suatu daftar?

mylist = ["apple", "banana", "cherry"]
print(type(mylist))
```

## Konstruktor list()
Dimungkinkan juga untuk menggunakan konstruktor list() saat membuat daftar baru.

```python
#Contoh
#Menggunakan list()konstruktor untuk membuat Daftar:

thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)
```

## Akses Item
Item daftar diindeks dan Anda dapat mengaksesnya dengan merujuk ke nomor indeks:

```python
#Contoh
#Cetak item kedua dari daftar:

thislist = ["apple", "banana", "cherry"]
print(thislist[1])
```

## Pengindeksan Negatif
Pengindeksan negatif berarti mulai dari akhir

-1merujuk pada item terakhir, -2merujuk pada item kedua terakhir, dst.

```python
#Contoh
#Cetak item terakhir dari daftar:

thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

## Rentang Indeks
Anda dapat menentukan rentang indeks dengan menentukan di mana memulai dan di mana mengakhiri rentang.

Saat menentukan rentang, nilai yang dikembalikan akan berupa daftar baru dengan item yang ditentukan.

```python
#Contoh
#Kembalikan item ketiga, keempat, dan kelima:

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
```

Catatan: Pencarian akan dimulai pada indeks 2 (termasuk) dan berakhir pada indeks 5 (tidak termasuk).

Ingat bahwa item pertama memiliki indeks 0.

Dengan menghilangkan nilai awal, rentang akan dimulai pada item pertama:

```python
#Contoh
#Contoh ini mengembalikan item dari awal ke, tetapi TIDAK termasuk, "kiwi":

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])
```

Dengan menghilangkan nilai akhir, rentang akan berlanjut hingga akhir daftar:

```python
#Contoh
#Contoh ini mengembalikan item dari "cherry" hingga akhir:

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
```

## Rentang Indeks Negatif
Tentukan indeks negatif jika Anda ingin memulai pencarian dari akhir daftar:

```python
#Contoh
#Contoh ini mengembalikan item dari "orange" (-4) hingga, tetapi TIDAK termasuk "mango" (-1):

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

## Periksa apakah Item Ada
Untuk menentukan apakah item tertentu ada dalam daftar, gunakan inkata kunci:

```python
#Contoh
#Periksa apakah "apple" ada dalam daftar:

thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("Yes, 'apple' is in the fruits list")
```

## Ubah Nilai Item
Untuk mengubah nilai item tertentu, lihat nomor indeks:

```python
#Contoh
#Ubah item kedua:

thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

## Mengubah Rentang Nilai Item
Untuk mengubah nilai item dalam rentang tertentu, tentukan daftar dengan nilai baru, dan rujuk rentang nomor indeks tempat Anda ingin memasukkan nilai baru:

```python
#Contoh
#Ubah nilai "pisang" dan "ceri" dengan nilai "kismis hitam" dan "semangka":

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)
```

Jika Anda memasukkan lebih banyak item daripada yang Anda ganti, item baru akan dimasukkan di tempat yang Anda tentukan, dan item yang tersisa akan dipindahkan sesuai dengan itu:

```python
#Contoh
#Ubah nilai kedua dengan menggantinya dengan dua nilai baru:

thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)
```

Catatan: Panjang daftar akan berubah jika jumlah item yang dimasukkan tidak sesuai dengan jumlah item yang diganti.

Jika Anda memasukkan item yang lebih sedikit daripada yang Anda ganti, item baru akan dimasukkan di tempat yang Anda tentukan, dan item yang tersisa akan dipindahkan sesuai dengan itu:

```python
#Contoh
#Ubah nilai kedua dan ketiga dengan menggantinya dengan satu nilai:

thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)
```

## Tambahkan Item
Untuk menambahkan item ke akhir daftar, gunakan append() metode:

```python
#Contoh
#Menggunakan append()metode untuk menambahkan item:

thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

## Sisipkan Item
Untuk menyisipkan item daftar baru, tanpa mengganti nilai apa pun yang ada, kita dapat menggunakan insert()metode tersebut.

Metode ini insert()menyisipkan item pada indeks yang ditentukan:

```python
#Contoh
#Masukkan "semangka" sebagai item ketiga:

thislist = ["apple", "banana", "cherry"]
thislist.insert(2, "watermelon")
print(thislist)
```

## Perluas Daftar
Untuk menambahkan elemen dari daftar lain ke daftar saat ini, gunakan extend()metode.

```python
#Contoh
#Tambahkan elemen tropicalke thislist:

thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)
```

Elemen-elemen akan ditambahkan ke akhir daftar.

## Tambahkan Iterable Apa Pun
Metode ini extend()tidak harus menambahkan daftar , Anda dapat menambahkan objek apa pun yang dapat diulang (tupel, set, kamus, dll.).

```python
#Contoh
#Tambahkan elemen tupel ke daftar:

thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)
```

## Hapus Item Tertentu
Metode ini remove()menghapus item yang ditentukan.

```python
#Contoh
#Hapus "pisang":

thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)
```

Jika ada lebih dari satu item dengan nilai yang ditentukan, metode remove()akan menghapus kemunculan pertama:

```python
#Contoh
#Hapus kemunculan pertama "banana":

thislist = ["apple", "banana", "cherry", "banana", "kiwi"]
thislist.remove("banana")
print(thislist)
```

## Hapus Indeks Tertentu
Metode ini pop()menghapus indeks yang ditentukan.

```python
#Contoh
#Hapus item kedua:

thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)
```

Jika Anda tidak menentukan indeks, pop()metode akan menghapus item terakhir.

```python
#Contoh
#Hapus item terakhir:

thislist = ["apple", "banana", "cherry"]
thislist.pop()
print(thislist)
```

Kata del kunci juga menghapus indeks yang ditentukan:

```python
#Contoh
#Hapus item pertama:

thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(thislist)
```

Kata del kunci juga dapat menghapus daftar sepenuhnya.

```python
#Contoh
#Hapus seluruh daftar:

thislist = ["apple", "banana", "cherry"]
del thislist
print(thislist)
```

## Hapus Daftar
Metode ini clear()mengosongkan daftar.

Daftarnya masih ada, tetapi tidak ada isinya.

```python
#Contoh
#Hapus konten daftar:

thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)
```

## Mengulang Daftar
Anda dapat melakukan pengulangan pada item-item dalam daftar dengan menggunakan for loop:

```python
#Contoh
#Cetak semua item dalam daftar, satu per satu:

thislist = ["apple", "banana", "cherry"]
for x in thislist:
  print(x)
```

Pelajari lebih lanjut tentang forloop di Bab Python For Loops kami.

## Ulangi Melalui Angka Indeks
Anda juga dapat melakukan pengulangan pada item-item daftar dengan mengacu pada nomor indeksnya.

Gunakan fungsi range()dan len()untuk membuat iterasi yang sesuai.

```python
#Contoh
#Cetak semua item dengan merujuk pada nomor indeksnya:

thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
  print(thislist[i])

#Iterable yang dibuat dalam contoh di atas adalah [0, 1, 2].
```

## Menggunakan Perulangan While
Anda dapat melakukan pengulangan pada item-item dalam daftar dengan menggunakan whileloop.

Gunakan len()fungsi tersebut untuk menentukan panjang daftar, lalu mulai dari 0 dan lakukan pengulangan melalui item daftar dengan merujuk pada indeksnya.

Ingatlah untuk menambah indeks sebesar 1 setelah setiap iterasi.

```python
#Contoh
#Cetak semua item, menggunakan whileloop untuk menelusuri semua nomor indeks

thislist = ["apple", "banana", "cherry"]
i = 0
while i < len(thislist):
  print(thislist[i])
  i = i + 1
```

Pelajari lebih lanjut tentang whileperulangan di Bab Perulangan While Python kami .

## Perulangan Menggunakan Pemahaman Daftar
Pemahaman Daftar menawarkan sintaksis terpendek untuk melakukan pengulangan melalui daftar:

```python
#Contoh
#Loop tangan pendek foryang akan mencetak semua item dalam daftar:

thislist = ["apple", "banana", "cherry"]
[print(x) for x in thislist]
```

## Urutkan Daftar Secara Alfanumerik
Objek daftar memiliki sort()metode yang akan mengurutkan daftar secara alfanumerik, menaik, secara default:

```python
#Contoh
#Urutkan daftar berdasarkan abjad:

thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort()
print(thislist)
```

```python
#Contoh
#Urutkan daftar secara numerik:

thislist = [100, 50, 65, 82, 23]
thislist.sort()
print(thislist)
```

## Urutkan Menurun
Untuk mengurutkan secara menurun, gunakan argumen kata kunci reverse = True:

```python
#Contoh
#Urutkan daftar secara menurun:

thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

```python
#Contoh
#Urutkan daftar secara menurun:

thislist = [100, 50, 65, 82, 23]
thislist.sort(reverse = True)
print(thislist)
```

## Sesuaikan Fungsi Sortir
Anda juga dapat menyesuaikan fungsi Anda sendiri dengan menggunakan argumen kata kunci .key = function

Fungsi ini akan mengembalikan angka yang akan digunakan untuk mengurutkan daftar (angka terendah terlebih dahulu):

```python
#Contoh
#Urutkan daftar berdasarkan seberapa dekat angka tersebut dengan 50:

def myfunc(n):
  return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```

## Pengurutan Tanpa Membedakan Huruf Besar/Kecil
Secara default, sort()metode ini peka huruf besar-kecil, sehingga semua huruf kapital diurutkan sebelum huruf kecil:

```python
#Contoh
#Pengurutan peka huruf besar/kecil dapat memberikan hasil yang tidak diharapkan:

thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort()
print(thislist)
```

Untungnya kita dapat menggunakan fungsi bawaan sebagai fungsi utama saat mengurutkan daftar.

Jadi jika Anda menginginkan fungsi pengurutan yang tidak peka huruf besar/kecil, gunakan str.lower sebagai fungsi kunci:

```python
#Contoh
#Lakukan pengurutan daftar tanpa memperhatikan huruf besar/kecil:

thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
```

## Urutan Terbalik
Bagaimana jika Anda ingin membalik urutan daftar, apa pun alfabetnya?

Metode ini reverse()membalik susunan penyortiran elemen saat ini.

```python
#Contoh
#Membalikkan urutan item daftar:

thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

## Salin Daftar
Anda tidak dapat menyalin daftar hanya dengan mengetik list2 = list1, karena: list2hanya akan menjadi referensi ke list1, dan perubahan yang dibuat di list1akan secara otomatis juga dibuat di list2.

## Gunakan metode copy()
Anda dapat menggunakan metode Daftar bawaan copy()untuk menyalin daftar.

```python
#Contoh
#Buat salinan daftar dengan copy()metode:

thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
```

## Gunakan metode list()
Cara lain untuk membuat salinan adalah dengan menggunakan metode bawaan list().

```python
#Contoh
#Buat salinan daftar dengan list()metode:

thislist = ["apple", "banana", "cherry"]
mylist = list(thislist)
print(mylist)
```

## Gunakan Operator irisan
Anda juga dapat membuat salinan daftar dengan menggunakan :operator (irisan).

```python
#Contoh
#Buat salinan daftar dengan :operator:

thislist = ["apple", "banana", "cherry"]
mylist = thislist[:]
print(mylist)
```

## Gabungkan Dua Daftar
Ada beberapa cara untuk menggabungkan, atau menyatukan, dua atau lebih daftar dalam Python.

Salah satu cara termudah adalah dengan menggunakan + operator.

```python
#Contoh
#Bergabunglah dengan dua daftar:

list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
```

Cara lain untuk menggabungkan dua daftar adalah dengan menambahkan semua item dari daftar2 ke daftar1, satu per satu:

```python
#Contoh
#Tambahkan list2 ke list1:

list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

for x in list2:
  list1.append(x)

print(list1)
```

Atau Anda dapat menggunakan extend() metode ini, yang tujuannya adalah untuk menambahkan elemen dari satu daftar ke daftar lainnya:

```python
#Contoh
#Gunakan extend()metode untuk menambahkan list2 di akhir list1:

list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list1.extend(list2)
print(list1)
```
