### 1. Buat fungsi hitung_vokal(teks) yang mengembalikan jumlah huruf vokal (a, i, u, e, o) dalam teks yang diberikan
#### a. Return Value
- Input: hitung_vokal("Saya Belajar AI")
- Output: 7
#### b. Tanpa Return Value
- Input: cetak_jumlah_vokal("Halo Dunia")
- Output: Jumlah vokal: 5
#### c. Passing parameter tambahan hitung_vokal_opsi(teks, ignore_case)
- Input: hitung_vokal_opsi("BelAjar PYthon", True)
- Output: 5

```python
# a
def hitung_vokal(teks):
    jumlah = 0
    for i in teks.lower():
        if i in ("a", "i", "u", "e", "o"):
            jumlah += 1
    return jumlah

teks = input("Masukkan teks: ")
hitung_vokal(teks)
```

```python
# b
def cetak_jumlah_vokal(teks):
    jumlah = 0
    for i in teks.lower():
        if i in ("a", "i", "u", "e", "o"):
            jumlah += 1
    print(f"Jumlah Vokal: {jumlah}")

teks = input("Masukkan teks: ")
cetak_jumlah_vokal(teks)
```

```python
# c
def hitung_vokal_opsi(teks, ignore_case):
    if ignore_case == "True":
        jumlah = 0
        for i in teks.lower():
            if i in ("a", "i", "u", "e", "o"):
                jumlah += 1
        print(f"Jumlah Vokal: {jumlah}")
    else:
        print(f"Ignore Case = {ignore_case}")
        print(f"Teks = {teks}")
        
teks = input("Masukkan teks: ")
case = input("Masukkan ignore case: ")
hitung_vokal_opsi(teks, case)
```

### 2. Buat fungsi rekursif hapus_duplikasi(teks) sehingga: "aabbccddaa" -> "abcd" (setiap karakter hanya muncul pertama kali)
- Input: "aabbccddaabb"
- Output: "abcd"

```python
def hapus_duplikasi(teks):
  if len(teks) == 1:
    return teks[0]
  else:
    duplikasi = hapus_duplikasi(teks[1:])
    return teks[0] if teks[0] > duplikasi else duplikasi

teks = input("Masukkan Teks: ")
hapus_duplikasi(teks)
```

### 3. Diberikan dictionary stok = {"buku":10, "pulpen":5, "penghapus":3}. Buat fungsi untuk memperbarui stok berdasarkan input barang dan jumlah penambahan/pengurangan.
- Input: Stok awal -> {"buku":10,"pulpen":5}
         Perintah -> ("buku", -3)
- Output: {"buku":7, "pulpen":5}

```python
def memperbarui(barang, jumlah, stok):
    stok[barang] += jumlah
    return stok

stok = {"buku":10, "pulpen":5, "penghapus":3}
nama = input("Masukkan Nama Barang: ")
perbarui = int(input("Masukkan Jumlah yang ingin diperbarui: "))
print(f"Stok awal: {stok}")
print(f"Stok setelah diperbarui: {memperbarui(nama, perbarui, stok)}")
```
