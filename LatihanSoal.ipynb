3. Rancang program untuk menampilkan deret bilangan ganjil sampai n dan
jumlahnya.

```python
n = int(input("Masukkan nilai n: "))

jumlah = 0

print("Deret bilangan ganjil sampai", n, ":")

for i in range(1, n+1, 2):
    print(i, end=" ")
    jumlah += i

print("\nJumlah deret bilangan ganjil sampai", n, "adalah:", jumlah)
```

## GAME SEDERHANA
1. Game kertas batu gunting

```python
import random
skor_u = 0
skor_c = 0
i = 0
while i != 5:
    print("1.Batu")
    print("2.Gunting")
    print("3.Kertas")
    user = int(input())
    comp = random.randint(1,3)
    print(comp)
    if (user == 1 and comp == 2) or (user == 2 and comp == 3) or (user == 3 and comp == 1) :
        print("Anda Menang")
        skor_u += 1
    elif (user == 1 and comp == 3) or (user == 2 and comp == 1) or (user == 3 and comp == 2) :
        print("Anda Kalah")
        skor_c += 1
    elif (user == comp):
        print("Seri")
    print(skor_u, skor_c)
    i += 1
```

2. tebak angka rahasia

```python
import random
nyawa = 5
x = random.randint(1,100)
while nyawa != 0:
    y = int(input("Masukkan Angka: "))
    if y > x:
        print("Lebih Kecil")
        nyawa -= 1
    elif y < x:
        print("Lebih Besar")
        nyawa -= 1
    else :
        print("Anda menang dengan", nyawa ,"nyawa tersisa")
        break
else :
    print("Game Over")
    print("Angka Rahasianya adalah", x )
```

# 3.Buat code untuk mengkonversi waktu dengan inputan berupa waktu dalam bentuk detik, dan output dalam bentuk jam, menit, serta detik, seperti contoh tampilan pada Gambar 5:
Screenshot 2025-09-24 192020.png

```python
detik = int(input("masukkan data waktu dalam detik: "))

jam = detik//3600
sisa_waktu = detik%3600
menit = sisa_waktu//60
sisa_detik = sisa_waktu%60

print(f"{detik} sama dengan {jam} jam, {menit} menit, {sisa_detik} detik")
```

6. Buatkan kode program yang meminta kepada pengguna untuk memasukkan suatu
integer dan periksa apkah bilangan tersebut dapat dibagi 3 dan 7, dapat dibagi 3 atau 7,
atau tidak dapat dibagi oleh keduanya!
Contoh tampilan saat program dijalankan:
Screenshot 2025-09-12 054012.png

```python
# 6.Jawaban: kode program
# Meminta input dari pengguna
num = int(input("Masukkan bilangan: "))

# Memeriksa kondisi pembagian
if num % 3 == 0 and num % 7 == 0:
    print(f"{num} dapat dibagi oleh 3 dan 7.")
elif num % 3 == 0 or num % 7 == 0:
    print(f"{num} dapat dibagi oleh 3 atau 7.")
else:
    print(f"{num} tidak dapat dibagi oleh 3 maupun 7.")
```

```python
# 3.Jawaban: code program
# Program konversi suhu dari Celcius ke Reamur, Fahrenheit, Rankine, dan Kelvin

# Input suhu dalam Celcius dari user
celcius = float(input("Masukkan suhu Celcius: "))

# Konversi ke Reamur
reamur = (4/5) * celcius

# Konversi ke Fahrenheit
fahrenheit = (9/5) * celcius + 32

# Konversi ke Rankine
rankine = (9/5) * celcius + 491.67

# Konversi ke Kelvin
kelvin = celcius + 273.15

# Menampilkan hasil konversi
print(f"Konversi suhu celcius ke suhu Reamur: {reamur:.2f} °R")
print(f"Konversi suhu celcius ke suhu Fahrenheit: {fahrenheit:.2f} °F")
print(f"Konversi suhu celcius ke suhu Rankine: {rankine:.2f} °Ra")
print(f"Konversi suhu celcius ke suhu Kelvin: {kelvin:.2f} K")
```
