##### Dictionary
dictionary digunakan untuk menyimpan nilai data dalam pasangan kunci:nilai.

dictionary adalah suatu koleksi yang teratur*, dapat diubah dan tidak memperbolehkan adanya duplikat.

Mulai dari Python versi 3.7, dictionary diurutkan . Pada Python 3.6 dan versi sebelumnya, dictionary tidak diurutkan .

dictionary Item tersusun rapi, dapat diubah, dan tidak memperbolehkan duplikat.

#### - Loop Dictionaries
##### Loop Melalui Kamus
Anda dapat mengulang kamus dengan menggunakan forloop.

for x in thisdict:

for x in thisdict.values():

for x in thisdict.keys():
  print(x)

Lakukan pengulangan pada kunci dan nilai , dengan menggunakan items()metode:

#### - Copy Dictionaries
##### Salin Kamus
Anda tidak dapat menyalin kamus hanya dengan mengetik dict2 = dict1, karena: dict2hanya akan menjadi referensi ke dict1, dan perubahan yang dibuat dalam dict1akan secara otomatis juga dibuat dalam dict2.

Contoh:

Cara lain untuk membuat salinan adalah dengan menggunakan fungsi bawaan dict().

#### - Dictionary Methods
##### Dictionary Methods
Python has a set of built-in methods that you can use on dictionaries.

clear() = Removes all the elements from the dictionary

copy() = Returns a copy of the dictionary

fromkeys() = Returns a dictionary with the specified keys and value

get() = Returns the value of the specified key

items() = Returns a list containing a tuple for each key value pair

keys() = Returns a list containing the dictionary's keys

pop() = Removes the element with the specified key

popitem() = Removes the last inserted key-value pair

setdefault() = Returns the value of the specified key. If the key does not exist: insert the key, with the specified value

update() = Updates the dictionary with the specified key-value pairs

values() = Returns a list of all the values in the dictionary

#### 2 skor_tugas = {"Dina": 90, "Roni": 75, "Lina": 88, "Tono": 82}
#### a. Hitung total skor seluruh siswa
#### b. Tampilkan nilai terendah
#### c. Hapus data Roni

```python
skor_tugas = {"Dina": 90, "Roni": 75, "Lina": 88, "Tono": 82}

print("a :")
total = 0
for v in skor_tugas.values():
    total += v
print("Total skor seluruh siswa =", total)

print("")
print("b :")
maks = 100
for v in skor_tugas.values():
    if v < maks:
        maks = v
print("Nilai terendah =", maks)

print("")
print("c :")
skor_tugas.pop("Roni")
print(skor_tugas)
```

```python
buku = [
{"judul": "Pemrograman Python", "kode": "BK001"},
{"judul": "Dasar AI", "kode": "BK002"},
{"judul": "Data Mining", "kode": "BK003"}
]

print("a :")
buku.append({"judul": "Introduction Python", "kode": "BK004"})
print(buku)

print("")
print("b :")
for item in buku:
    print(f"Judul: {item['judul']}, Kode: {item['kode']}")

print("")
print("c :")
cari = input("Masukkan kode buku yang ingin di cari: ")
for value in buku:
    if cari == value["kode"]:
        print("Judul =", value["judul"])

kode = input("Masukkan kode buku yang ingin di cari: ")
for item in buku:
    if kode not in item["kode"]:
        print("Buku tidak tersedia")
        break
```

#### 5. pesanan = {"Nasi Goreng": (15000, 2), "Mie Ayam": (12000, 1), "Sate": (20000, 3)}
#### a. Hitung total pendapatan dari semua pesanan
#### b. Tampilkan menu dengan total pembayaran terbesar
####  (harga × jumlah)

```python
pesanan = {"Nasi Goreng": (15000, 2), "Mie Ayam": (12000, 1), "Sate": (20000, 3)}

print("a :")
total = 0
for harga, jumlah in pesanan.values():
    total += harga * jumlah
print("Total pendapatan:", total)

print("")
print("b :")

menu = ""
pembayaran_terbesar = 0
for nama, (harga, jumlah) in pesanan.items():
    total_menu = harga * jumlah
    if total_menu > pembayaran_terbesar:
        pembayaran_terbesar = total_menu
        menu = nama
print("Menu dengan total pembayaran terbesar:", menu, "=", pembayaran_terbesar)
```
