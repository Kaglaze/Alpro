```python
# 15. buatlah fungsi bernama kelilingpersegi(s) yang mengembalikan keliling persegi
def kelilingpersegi(s):
    keliling = 4 * s
    return keliling

s = int(input("Masukkan Keliling: "))
kelilingpersegi(s)
```

```python
# 16. Buat fungsi luas segitiga(a,t) untuk menghitung luas segitiga
def luassegitiga(a,t):
    luas = (1/2) * a * t
    return luas

a = int(input("Masukkan alas: "))
t = int(input("Masukkan tinggi: "))
luassegitiga(a,t)
```

```python
# 17
def cekgenap(x):
    bolean = False
    if x % 2 == 0:
        bolean = True
        return bolean
    else:
        return bolean

x = int(input("Masukkan angka: "))
cekgenap(x)
```

```python
# 19
def ratarata(listangka):
    a = 0
    l = len(listangka)
    for i in listangka:
        a += i
    total = a / l
    print(total)

listangka = [1, 2, 3, 4, 5]
ratarata(listangka)
```
